```python
import requests
import json
from copy import deepcopy

GLASSDOOR_URL = "https://www.glassdoor.com/graph"


def fetch_all_companies(
    headers: dict,
    base_payload: list,
    start_page: int = 1,
    min_results_per_page: int = 10,
    output_file: str = "companies.json",
):
    """
    Loop through Glassdoor employer search pages until results < min_results_per_page
    and store all company data into one JSON file.
    """

    all_companies = []
    current_page = start_page

    while True:
        payload = deepcopy(base_payload)
        payload[0]["variables"]["employerSearchInput"]["pageRequested"] = current_page

        print(f"Fetching page {current_page}...")

        response = requests.post(GLASSDOOR_URL, headers=headers, json=payload)
        response.raise_for_status()

        data = response.json()
        results = (
            data[0]
            .get("data", {})
            .get("employerSearchRG", {})
            .get("employerResults", [])
        )

        # Stop condition
        if len(results) < min_results_per_page:
            print("Result count < minimum threshold. Stopping.")
            break

        for item in results:
            employer = item.get("employer", {})
            ratings = employer.get("ratings", {})

            company_data = {
                "company_id": employer.get("id"),
                "name": employer.get("shortName"),
                "profile_id": employer.get("bestProfile", {}).get("id"),
                "ratings": {
                    "overall": ratings.get("overallRating"),
                    "career_opportunities": ratings.get("careerOpportunitiesRating"),
                    "compensation_benefits": ratings.get("compensationAndBenefitsRating"),
                    "culture_values": ratings.get("cultureAndValuesRating"),
                    "diversity_inclusion": ratings.get("diversityAndInclusionRating"),
                    "senior_management": ratings.get("seniorManagementRating"),
                    "work_life_balance": ratings.get("workLifeBalanceRating"),
                },
                "demographic_ratings": item.get("demographicRatings", []),
                "best_places_to_work": employer.get("bestPlacesToWork", []),
                "best_led_companies": employer.get("bestLedCompanies", []),
            }

            all_companies.append(company_data)

        current_page += 1

    # Save to JSON
    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(all_companies, f, indent=2, ensure_ascii=False)

    print(f"Saved {len(all_companies)} companies to {output_file}")
    return all_companies
```

```python
import requests
import json
import re
from copy import deepcopy

GLASSDOOR_URL = "https://www.glassdoor.com/graph"


def slugify_company_name(name: str) -> str:
    """
    Convert company name into Glassdoor-friendly slug
    Example: 'Helping Hands Home Care' -> 'Helping-Hands-Home-Care'
    """
    name = re.sub(r"[^\w\s-]", "", name)
    return re.sub(r"\s+", "-", name).strip("-")


def build_company_url(company_name: str, company_id: int) -> str:
    slug = slugify_company_name(company_name)
    return (
        f"https://www.glassdoor.com/Overview/"
        f"Working-at-{slug}-EI_IE{company_id}.11,16.htm"
    )


def fetch_all_companies(
    headers: dict,
    base_payload: list,
    start_page: int = 1,
    min_results_per_page: int = 10,
    output_file: str = "companies.json",
):
    all_companies = []
    current_page = start_page

    while True:
        payload = deepcopy(base_payload)
        payload[0]["variables"]["employerSearchInput"]["pageRequested"] = current_page

        print(f"Fetching page {current_page}...")

        response = requests.post(GLASSDOOR_URL, headers=headers, json=payload)
        response.raise_for_status()

        data = response.json()
        results = (
            data[0]
            .get("data", {})
            .get("employerSearchRG", {})
            .get("employerResults", [])
        )

        if len(results) < min_results_per_page:
            print("Result count < threshold. Stopping.")
            break

        for item in results:
            employer = item.get("employer", {})
            ratings = employer.get("ratings", {})

            company_id = employer.get("id")
            company_name = employer.get("shortName")

            company_data = {
                "company_id": company_id,
                "name": company_name,
                "company_url": build_company_url(company_name, company_id),
                "profile_id": employer.get("bestProfile", {}).get("id"),
                "ratings": {
                    "overall": ratings.get("overallRating"),
                    "career_opportunities": ratings.get("careerOpportunitiesRating"),
                    "compensation_benefits": ratings.get("compensationAndBenefitsRating"),
                    "culture_values": ratings.get("cultureAndValuesRating"),
                    "diversity_inclusion": ratings.get("diversityAndInclusionRating"),
                    "senior_management": ratings.get("seniorManagementRating"),
                    "work_life_balance": ratings.get("workLifeBalanceRating"),
                },
                "demographic_ratings": item.get("demographicRatings", []),
                "best_places_to_work": employer.get("bestPlacesToWork", []),
                "best_led_companies": employer.get("bestLedCompanies", []),
            }

            all_companies.append(company_data)

        current_page += 1

    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(all_companies, f, indent=2, ensure_ascii=False)

    print(f"Saved {len(all_companies)} companies to {output_file}")
    return all_companies
```
