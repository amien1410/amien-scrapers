```python
# Drop useless column
df = df.drop(columns=['product_url'])

# Convert datetime columns
df['sale_starts_at'] = pd.to_datetime(df['sale_starts_at'])
df['sale_ends_at'] = pd.to_datetime(df['sale_ends_at'])

df.head()
```

```python
df['width_depth'] = df['width'] * df['depth']
df['width_height'] = df['width'] * df['height']
df['duration'] = (df['sale_ends_at'] - df['sale_starts_at']).dt.total_seconds()
```

```python
df['manufacturer_clean'] = df['manufacturer'].str.lower()

df['manufacturer_clean'].value_counts()
```
