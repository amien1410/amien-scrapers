```python
#@title Google Places API
import sys
# !{sys.executable} -m pip install google-maps-places

from google.maps import places_v1
from google.geo.type.types import Viewport
from google.maps.places_v1 import types as places_v1_types
from google.type import latlng_pb2
from google.protobuf.json_format import MessageToDict, MessageToJson
import pandas as pd
import json
import math

def get_places(API_KEY):
    client = places_v1.PlacesClient(
        client_options={"api_key": API_KEY}
    )

    # The address or place name to search for
    address = "Starbucks Singapore"

    # Create a SearchTextRequest object
    # Note: The 'fields' parameter is NOT included here.
    request = places_v1.SearchTextRequest(text_query=address)
    # Include the field mask
    list_of_fields = ["id","displayName", "formattedAddress", "primaryType", "types", "priceLevel","priceRange", "rating", "userRatingCount", "location"]
    # fieldMask = "places.id,places.formattedAddress,places.displayName"
    fieldMask = ""
    for x in range(len(list_of_fields)):
        fieldMask += "places."
        fieldMask += f"{list_of_fields[x]}"
        if x != len(list_of_fields) - 1:
            fieldMask += ","

    response = client.search_text(request=request, metadata=[("x-goog-fieldmask",fieldMask)])
    return display(response)
```
