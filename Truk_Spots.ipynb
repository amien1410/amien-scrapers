```python
def combine_json_files(data_folder, output_file):
    combined_data = []

    # Traverse through each folder in the data directory
    for country_folder in os.listdir(data_folder):
        country_path = os.path.join(data_folder, country_folder)

        # Ensure it's a directory (for example, skip any non-folder files if any)
        if os.path.isdir(country_path):
            # Traverse each JSON file in the country folder
            for json_file in os.listdir(country_path):
                if json_file.endswith('.json'):
                    json_file_path = os.path.join(country_path, json_file)

                    # Load JSON data and append it to the combined_data list
                    with open(json_file_path, 'r', encoding='utf-8') as f:
                        data = json.load(f)
                        combined_data.append(data)

    # Write the combined data to the output JSON file
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(combined_data, f, indent=4)

    print(f"Combined JSON file created at: {output_file}")

# Example usage
combine_json_files('/content/ocm-export/data', 'raw_data.json')
```

```python
import json
import csv
from typing import List, Dict, Any

def flatten_json_to_csv(data: Dict[str, Any]) -> dict:
    """
    Flatten a nested JSON object into a single-level dictionary suitable for CSV conversion.
    Handles nested AddressInfo and multiple Connections.

    Args:
        data: Dictionary containing the charging station data

    Returns:
        Flattened dictionary with all nested structures converted to flat columns
    """
    flattened = {}

    # Process regular fields
    for key, value in data.items():
        if key not in ['AddressInfo', 'Connections']:
            flattened[key] = value

    # Process AddressInfo
    if 'AddressInfo' in data:
        for key, value in data['AddressInfo'].items():
            flattened[f'AddressInfo_{key}'] = value

    # Process Connections
    if 'Connections' in data:
        for i, connection in enumerate(data['Connections'], 1):
            for key, value in connection.items():
                flattened[f'Connection_{i}_{key}'] = value

    return flattened

def json_array_to_csv(input_data: List[Dict[str, Any]], output_file: str) -> None:
    """
    Convert an array of JSON data to CSV format and save to file.

    Args:
        input_data: List of dictionaries containing charging station data
        output_file: Path to the output CSV file
    """
    if not input_data:
        raise ValueError("Input data array is empty")

    # Flatten the first record to get all possible columns
    flattened_first = flatten_json_to_csv(input_data[0])
    all_fields = list(flattened_first.keys())

    # Collect all possible fields from all records
    for record in input_data[1:]:
        flattened_record = flatten_json_to_csv(record)
        for field in flattened_record.keys():
            if field not in all_fields:
                all_fields.append(field)

    # Write to CSV
    with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=all_fields)
        writer.writeheader()

        # Write each record
        for record in input_data:
            flattened_record = flatten_json_to_csv(record)
            # Ensure all fields are present (even if empty)
            for field in all_fields:
                if field not in flattened_record:
                    flattened_record[field] = None
            writer.writerow(flattened_record)
```
