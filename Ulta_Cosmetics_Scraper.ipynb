```python
def get_payload(page):
    return {
        "operationName": "NonCachedPage",
        "query": """
            query NonCachedPage($stagingHost: String, $previewOptions: JSON, $moduleParams: JSON) {
              Page: NonCachedPage(
                stagingHost: $stagingHost,
                previewOptions: $previewOptions,
                moduleParams: $moduleParams,
                url: {path: "https://www.ulta.com/shop/makeup/eyes?page=""" + str(page) + """"},
                contentId: "cb7c0efb-8772-4abc-9be0-4dfaf1b625ee"
              ) {
                content
                meta
                __typename
              }
            }
        """,
        "variables": {
            "moduleParams": {
                "gti": "c35d1f5d-4e9a-45aa-abb7-a4fd841cd224",
                "loginStatus": "anonymous",
                "retailerVisitorId": "438f1fdcd3f76e5b62442e132abb4253",
                "breakpoint": "MD"
            }
        }
    }

all_products = []
page = 1

while True:
    print(f"📄 Fetching page {page}...")
    response = requests.post(url, headers=headers, json=get_payload(page))

    if response.status_code != 200:
        print(f"❌ Failed to fetch page {page}, status code: {response.status_code}")
        break

    try:
        items = response.json()['data']['Page']['content']['items']
        if not items:
            print("✅ No more products found. Stopping.")
            break

        for item in items:
            try:
                data_layer = item['dataCapture']['dataLayer']['Tealium']
                all_products.append({
                    'brand': item.get('brandName', ''),
                    'product_name': item.get('productName', ''),
                    'price': item.get('listPrice', '').replace('$', ''),
                    'product_id': item.get('productId', ''),
                    'sku': item.get('skuId', ''),
                    'url': item['action'].get('url', ''),
                    'image_url': item['image'].get('imageUrl', '')
                    # 'filter_label': data_layer.get('filter_label', [''])[0]
                })
            except Exception as e:
                print(f"⚠️ Error processing item on page {page}: {e}")

        page += 1

    except Exception as e:
        print(f"❌ Error parsing response for page {page}: {e}")
        break

df = pd.DataFrame(all_products)
df
```
