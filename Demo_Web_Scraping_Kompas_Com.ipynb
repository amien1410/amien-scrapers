```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://search.kompas.com/search?q=khalid+basalamah&site_id=all&start_date=2025-08-01&end_date=2025-10-04&page=2"
res = requests.get(url)
res.text
```

```python
soup = BeautifulSoup(res.text)
soup.title
```

```python
articles = soup.find_all("div", class_="articleItem")
articles
```

```python
link = articles[0].find("a", class_="article-link")['href']
judul = articles[0].find("h2", class_="articleTitle").get_text(strip=True)
kategori = articles[0].find("div", class_="articlePost-subtitle").get_text(strip=True)
tanggal_berita = articles[0].find("div", class_="articlePost-date").get_text(strip=True)
```

```python
berita = []
for article in articles:
  berita.append({
      "judul": article.find("a", class_="article-link")['href'],
      "link": article.find("h2", class_="articleTitle").get_text(strip=True),
      "kategori": article.find("div", class_="articlePost-subtitle").get_text(strip=True),
      "tanggal_berita": article.find("div", class_="articlePost-date").get_text(strip=True)
  })
berita
```

```python
berita = []
queries = ["khalid basalamah", "yaqut cholil", "ibnu mas'ud", "korupsi kuota haji"]

for query in queries:
  page = 1
  while True:
    url = f"https://search.kompas.com/search?q={query}&site_id=all&start_date=2025-08-01&end_date=2025-10-04&page={page}"
    print(f"Sedang mencari {url}")
    res = requests.get(url)
    soup = BeautifulSoup(res.text)
    articles = soup.find_all("div", class_="articleItem")

    if len(articles) == 0:
      break

    for article in articles:
      berita.append({
          "query": query,
          "judul": article.find("a", class_="article-link")['href'],
          "link": article.find("h2", class_="articleTitle").get_text(strip=True),
          "kategori": article.find("div", class_="articlePost-subtitle").get_text(strip=True),
          "tanggal_berita": article.find("div", class_="articlePost-date").get_text(strip=True)
      })

    page += 1

df = pd.DataFrame(berita)
df
```

```python
df.to_csv("berita-korupsi-kuota-haji.csv")
```
