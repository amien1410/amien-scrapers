```python
import re

def extract_workflow_details(workflows_list):
    extracted_data = []
    for workflow in workflows_list:
        integrations_str = ", ".join(workflow.get("integrations", [])) if isinstance(workflow.get("integrations"), list) else ""

        download_url = workflow.get("download_url", "")
        file_location = ""
        # Extract the path from the download_url
        match = re.search(r"https://raw.githubusercontent.com/Zie619/n8n-workflows/main/(.*)", download_url)
        if match:
            file_location = match.group(1)
        else:
            # Fallback if the URL doesn't match the expected GitHub raw pattern
            file_location = download_url.split("https://raw.githubusercontent.com/Zie619/n8n-workflows/main/")[-1]

        extracted_data.append({
            "id": workflow.get("id"),
            "name": workflow.get("name"),
            "description": workflow.get("description"),
            "filename": workflow.get("filename"),
            "complexity": workflow.get("complexity"),
            "node_count": workflow.get("node_count"),
            "integrations": integrations_str,
            "category": workflow.get("category"),
            "file_location": file_location
        })
    return extracted_data


records = extract_workflow_details(data['workflows'])

# Print as CSV-like output
for r in records:
    print(f"{r['id']}, {r['name']}, \"{r['description']}\", {r['complexity']}, {r['node_count']}, {r['integrations']}, {r['category']}, {r['file_location']}")
```

```python
import re
import json

def parse_blocks(text):
    lines = [l.strip() for l in text.split("\n") if l.strip()]
    results = []
    app = None
    i = 0

    while i < len(lines):
        line = lines[i]

        # Detect app name (lowercase, no spaces)
        if re.fullmatch(r"[a-z]+", line):
            app = line
            i += 1
            continue

        # Title
        title = line

        # Description
        if i + 1 < len(lines) and lines[i+1].startswith("Description:"):
            description = lines[i+1].replace("Description:", "").strip()
        else:
            description = ""

        # Template
        if i + 2 < len(lines) and lines[i+2].startswith("Template:"):
            template = lines[i+2].replace("Template:", "").strip()
        else:
            template = ""

        if app:
            results.append({
                "appName": app,
                "title": title,
                "description": description,
                "templateUrl": template
            })

        i += 3  # jump to next block

    return results


records = parse_blocks(myText)

# Print as CSV-like output
for r in records:
    print(f"{r['appName']}, {r['title']}, \"{r['description']}\" , {r['templateUrl']}")
```

```python
import requests
import re
import os

# Create a folder for downloads
download_dir = "/content/downloads"
os.makedirs(download_dir, exist_ok=True)


def extract_file_id(url):
    """Extract Google Drive file ID."""
    match = re.search(r"id=([^&]+)", url)
    return match.group(1) if match else None


def get_original_filename(response, default_name="file"):
    """
    Extract original filename from Content-Disposition header only.
    Always ignore URL filename parameter.
    """
    cd = response.headers.get("Content-Disposition")

    if cd:
        # filename="example.json"
        fname_match = re.search(r'filename="(.+?)"', cd)
        if fname_match:
            return fname_match.group(1)

        # fallback pattern (sometimes Google uses filename*=UTF-8'')
        fname_utf_match = re.search(r"filename\*\=UTF-8''(.+)", cd)
        if fname_utf_match:
            return fname_utf_match.group(1)

    # If Google Drive does not supply filename, use ID
    return default_name


def download_file_from_drive(file_id, save_dir):
    """Download Google Drive file and preserve the original drive filename."""
    base_url = f"https://drive.google.com/uc?export=download&id={file_id}"
    session = requests.Session()

    # Step 1: initial request
    response = session.get(base_url, stream=True)

    # Check for confirmation token (large files)
    token = None
    for key, value in response.cookies.items():
        if key.startswith("download_warning"):
            token = value
            break

    # Step 2: re-download with confirmation token if needed
    if token:
        confirm_url = base_url + "&confirm=" + token
        response = session.get(confirm_url, stream=True)

    # Step 3: Real filename from Google's response
    original_name = get_original_filename(response, default_name=file_id)

    save_path = os.path.join(save_dir, original_name)

    # Step 4: Save the file
    with open(save_path, "wb") as f:
        for chunk in response.iter_content(1024):
            f.write(chunk)

    print(f"Downloaded  {original_name}")
    return original_name, save_path


# Process and update records
for record in records:
    url = record["templateUrl"]
    file_id = extract_file_id(url)

    if not file_id:
        print(f"Skipping invalid URL: {url}")
        continue

    filename, fullpath = download_file_from_drive(file_id, download_dir)

    # Add original filename to the record
    record["filename"] = filename
    record["filePath"] = fullpath


# Show updated records
print("\nUpdated Records:")
for r in records:
    print(r)
```
