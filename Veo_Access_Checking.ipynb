```python
from google import genai
from google.genai import errors

api_keys = [
]

def check_veo_access(api_key):
    try:
        client = genai.Client(api_key=api_key)
        # Attempt to list models to see if Veo is available for this key
        models = client.models.list()

        # Look for "veo" in the model names
        veo_models = [m.name for m in models if "veo" in m.name.lower()]

        if veo_models:
            print(f"[✅] SUCCESS: Key {api_key[:8]}... has access to: {veo_models}")
            return True
        else:
            print(f"[❌] NO ACCESS: Key {api_key[:8]}... cannot see Veo models.")
            return False

    except errors.ClientError as e:
        print(f"[⚠️] ERROR: Key {api_key[:8]}... is invalid or restricted: {e}")
        return False

print("Checking API keys for Google Veo access...\n")
valid_keys = []

for key in api_keys:
    if check_veo_access(key):
        valid_keys.append(key)

print(f"\nScan complete. {len(valid_keys)} key(s) found with Veo access.")
```
