```python
import json
import csv

def extract_company_data(data):
    results = []

    for item in data:
        phone = item.get("phone", "").replace(" ", "")
        address = " ".join(filter(None, [item.get("adresse"), item.get("adresse_compl_1")]))
        company = {
            "company_name": item.get("rs") or item.get("dc"),
            "company_id": item.get("id_company"),
            "address": address,
            "city": item.get("city"),
            "postcode": item.get("postcode"),
            "company_url": item.get("link"),
            "phone_number": phone,
            "website": item.get("website"),
        }
        results.append(company)

    return results


def save_to_csv(data, filename="companies.csv"):
    if not data:
        return

    keys = data[0].keys()

    with open(filename, mode="w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=keys)
        writer.writeheader()
        writer.writerows(data)


if __name__ == "__main__":
    extracted = extract_company_data(data)

    # Print result
    for row in extracted:
        print(row)

    # Save to CSV
    save_to_csv(extracted)
```

```python
import requests
import json
import time


BASE_URL = "https://www.gesec.fr/wp-json/sn/adherent/one"


# -----------------------------
# Extract basic company fields
# -----------------------------
def extract_basic(data):
    results = []

    for item in data:
        company = {
            "company_name": item.get("rs") or item.get("dc"),
            "company_id": item.get("id_company"),
            "address": item.get("adresse"),
            "city": item.get("city"),
            "postcode": item.get("postcode"),
            "company_url": item.get("link"),
            "phone_number": item.get("phone"),
            "website": item.get("website"),
        }
        results.append(company)

    return results


# -----------------------------
# Fetch detailed data
# -----------------------------
def fetch_details(company_id):
    url = f"{BASE_URL}?id_company={company_id}"

    try:
        response = requests.get(url, timeout=15)
        response.raise_for_status()
        data = response.json()
    except Exception as e:
        print(f"[ERROR] {company_id}: {e}")
        return None

    return extract_details(data)


# -----------------------------
# Extract only required fields
# -----------------------------
def extract_details(data):
    # Parse qualifications
    qualifications = []
    raw_qualifications = data.get("list_qualifications", {})

    if isinstance(raw_qualifications, dict):
        for value in raw_qualifications.values():
            if isinstance(value, dict):
                if value.get("label"):
                    qualifications.append(value["label"])
            elif isinstance(value, str):
                qualifications.append(value)

    return {
        "lat": data.get("lat"),
        "lng": data.get("lng"),
        "description": data.get("description"),
        "email": data.get("email"),
        "facebook": data.get("facebook"),
        "google": data.get("google"),
        "instagram": data.get("instagram"),
        "linkedin": data.get("linkedin"),
        "pinterest": data.get("pinterest"),
        "twitter": data.get("twitter"),
        "presentation_entreprise": data.get("presentation_entreprise"),
        "logo_entreprise": data.get("logo_entreprise"),
        "sav_catchment_kms": data.get("sav_catchment_kms"),
        "maintenance_catchment_kms": data.get("maintenance_catchment_kms"),
        "works_catchment_kms": data.get("works_catchment_kms"),
        "works_catchment_departments": data.get("works_catchment_departments"),
        "list_qualifications": qualifications,
    }


# -----------------------------
# Main pipeline
# -----------------------------
def enrich_companies(raw_data):
    # raw_data = extract_basic(raw_data)
    enriched = []

    for i, company in enumerate(raw_data, start=1):
        company_id = company.get("company_id")

        if not company_id:
            continue

        print(f"[{i}/{len(raw_data)}] Fetching {company_id}")

        details = fetch_details(company_id)

        if details:
            company.update(details)

        enriched.append(company)

        # Avoid rate limit
        time.sleep(0.3)

    return enriched


# -----------------------------
# Run script
# -----------------------------
if __name__ == "__main__":
    # Load your extracted JSON file
    # with open("input.json", "r", encoding="utf-8") as f:
    #     raw_data = json.load(f)

    results = enrich_companies(extracted)

    # Save final dataset
    with open("enriched_companies2.json", "w", encoding="utf-8") as f:
        json.dump(results, f, ensure_ascii=False, indent=2)

    print(f"\n✅ Done! Saved {len(results)} companies.")
```

```python
import re
from html import unescape
import json


# -----------------------------
# Clean HTML → readable text
# -----------------------------
def clean_html_text(raw_html):
    if not raw_html:
        return ""

    text = unescape(raw_html)

    # Replace separators
    text = text.replace("####", "\n\n").replace("##", "\n\n")

    # Remove HTML tags
    text = re.sub(r"<.*?>", "", text)

    # Normalize whitespace
    text = re.sub(r"\n\s*\n", "\n\n", text)
    text = re.sub(r"[ \t]+", " ", text)

    return text.strip()


# -----------------------------
# Convert list → string
# -----------------------------
def list_to_string(value):
    if not value:
        return ""

    if isinstance(value, list):
        return ", ".join(value)

    return str(value)


# -----------------------------
# Remove department codes
# "01 - Ain" → "Ain"
# -----------------------------
def clean_departments(value):
    if not value:
        return ""

    if isinstance(value, list):
        cleaned = []
        for item in value:
            # Remove leading "XX - "
            name = re.sub(r"^\d+\s*-\s*", "", item)
            cleaned.append(name)
        return ", ".join(cleaned)

    # fallback if already string
    return re.sub(r"\d+\s*-\s*", "", value)


# -----------------------------
# MAIN CLEANING FUNCTION
# -----------------------------
def clean_results(results):
    for company in results:
        # Clean text fields
        company["description"] = clean_html_text(company.get("description"))
        company["presentation_entreprise"] = clean_html_text(
            company.get("presentation_entreprise")
        )

        # Clean departments
        company["works_catchment_departments"] = clean_departments(
            company.get("works_catchment_departments")
        )

        # Clean qualifications
        company["list_qualifications"] = list_to_string(
            company.get("list_qualifications")
        )

    return results


# -----------------------------
# RUN CLEANING
# -----------------------------
if __name__ == "__main__":
    # Load your existing enriched file
    with open("enriched_companies2.json", "r", encoding="utf-8") as f:
        results = json.load(f)

    cleaned = clean_results(results)

    # Save cleaned version
    with open("cleaned_companies2.json", "w", encoding="utf-8") as f:
        json.dump(cleaned, f, ensure_ascii=False, indent=2)

    print(f"✅ Cleaned {len(cleaned)} companies")
```
