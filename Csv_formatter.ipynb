```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Upwork/mahmoud/april/moustaphalabban_products.csv')
display(df.head())
```

```python
df['product_url'] = 'https://moustaphalabban.com/products/' + df['handle']
display(df.head())
```

```python
df['public_variant_price'] = df['variant_price'].apply(lambda x: f'${x/100:.2f}')
display(df.head())
```

```python
df['public_variant_compare_at_price'] = df['variant_compare_at_price'].apply(lambda x: f'${x/100:.2f}')
display(df.head())
```

```python
output_filename = 'moustaphalabban_products_processed.csv'
df.to_csv(output_filename, index=False)
print(f"DataFrame successfully saved to {output_filename}")
```
