```python

def extract_api_keys(text):
    """
    Extracts API keys from a given text based on the pattern:
    alphanumeric string containing 'msh' and 'jsn'.
    """
    # Regex pattern to find alphanumeric strings containing 'msh' and 'jsn'
    # \b ensures whole word matching
    # [a-zA-Z0-9]+ matches one or more alphanumeric characters
    pattern = r'\b[a-zA-Z0-9]+msh[a-zA-Z0-9]+jsn[a-zA-Z0-9]+\b'
    api_keys = re.findall(pattern, text)
    return api_keys
```
