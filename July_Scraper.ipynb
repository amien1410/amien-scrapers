```python
extracted_data = []
for item in data:
    product = {
        "productName": item.get('display_name'),
        "Url": "https://www.maccosmetics.com" + item.get('product_url'),
        "Image": item.get('skus', {}).get('items', [{}])[0].get('media', {}).get('large', [{}])[0].get('src'),
        "Price": item.get('skus', {}).get('items', [{}])[0].get('prices', [{}])[0].get('include_tax', {}).get('price'),
        "Size": item.get('skus', {}).get('items', [{}])[0].get('sizes', [{}])[0].get('value'),
        "Ingredients": item.get('skus', {}).get('items', [{}])[0].get('iln_listing'),
        "description": item.get('meta', {}).get('description'),
    }
    extracted_data.append(product)

import pandas as pd

df = pd.DataFrame(extracted_data)
df
```

```python
def extract_and_pair_data(data_dict):
    """
    Extracts and pairs data from a dictionary where values are lists.

    Args:
        data_dict: A dictionary where each key's value is a list of the same length.

    Returns:
        A list of dictionaries, where each dictionary contains paired elements
        from the input dictionary's lists. Returns an empty list if the input
        dictionary is empty or lists have different lengths.
    """
    if not data_dict:
        return []

    # Get the keys and the lists of values
    keys = list(data_dict.keys())
    value_lists = list(data_dict.values())

    # Check if all lists have the same length
    it = iter(value_lists)
    the_len = len(next(it))
    if not all(len(l) == the_len for l in it):
        print("Error: Lists in the dictionary have different lengths.")
        return []

    extracted_data = []
    for i in range(the_len):
        record = {}
        for j, key in enumerate(keys):
            record[key] = value_lists[j][i]
        extracted_data.append(record)

    return extracted_data

# paired_data = extract_and_pair_data(eye_makeup)
# print(len(paired_data))
```

```python
ulta_raw = pd.read_csv("ulta_makeup_combined_size_products.csv")
ulta_raw
```

```python
ulta_raw.describe()
```

```python
nan_count = ulta_raw['variant_size'].isnull().sum()
print(f"The number of NaN values in the 'variant_size' column is: {nan_count}")
```

```python
ulta_raw['is_vegan'] = "no"
ulta_raw['is_cruelty_free'] = "no"
ulta_raw['is_fragrance_free'] = "no"
```

```python
lips_vegan_df = pd.read_csv('/content/ulta_makeup_lips_vegan.csv')
print(f"The number of products labeled as 'vegan' is: {lips_vegan_df.shape[0]}")

lips_cruelty_free_df = pd.read_csv('/content/ulta_makeup_lips_cruelty_free.csv')
print(f"The number of products labeled as 'cruelty_free' is: {lips_cruelty_free_df.shape[0]}")

lips_fragrance_free_df = pd.read_csv('/content/ulta_makeup_lips_fragrance_free.csv')
print(f"The number of products labeled as 'fragrance_free' is: {lips_fragrance_free_df.shape[0]}")
```

```python
ulta_raw.loc[ulta_raw['product_id'].isin(lips_vegan_df['product_id']), 'is_vegan'] = 'yes'
ulta_raw.loc[ulta_raw['product_id'].isin(lips_cruelty_free_df['product_id']), 'is_cruelty_free'] = 'yes'
ulta_raw.loc[ulta_raw['product_id'].isin(lips_fragrance_free_df['product_id']), 'is_fragrance_free'] = 'yes'
```

```python
vegan_count = ulta_raw[ulta_raw['is_vegan'] == 'yes'].shape[0]
print(f"The number of products marked as 'yes' in the 'is_vegan' column is: {vegan_count}")

cruelty_free_count = ulta_raw[ulta_raw['is_cruelty_free'] == 'yes'].shape[0]
print(f"The number of products marked as 'yes' in the 'is_cruelty_free' column is: {cruelty_free_count}")

fragrance_free_count = ulta_raw[ulta_raw['is_fragrance_free'] == 'yes'].shape[0]
print(f"The number of products marked as 'yes' in the 'is_fragrance_free' column is: {fragrance_free_count}")
```
