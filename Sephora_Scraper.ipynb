```python
import pandas as pd
from bs4 import BeautifulSoup
import httpx
import asyncio

# Load product URLs
df = pd.read_csv('/content/all_products.csv')

# Limit concurrent requests
semaphore = asyncio.Semaphore(5)  # You can adjust this between 5–20

async def get_product_size(client, url):
    async with semaphore:
        try:
            response = await client.get(url, headers=BASE_HEADERS)
            soup = BeautifulSoup(response.content, 'html.parser')
            size_span = soup.find('span', {'data-at': 'sku_size_label'})
            if size_span:
                return size_span.get_text(strip=True)
        except Exception as e:
            print(f"Error fetching {url}: {e}")
        return "Not Found"  # fallback when error or not found

async def fetch_all_sizes(df):
    async with httpx.AsyncClient(
        http2=True,
        headers=BASE_HEADERS,
        timeout=httpx.Timeout(15.0),
        limits=httpx.Limits(max_connections=5)
    ) as client:
        tasks = [get_product_size(client, url) for url in df['Product URL']]
        sizes = await asyncio.gather(*tasks)
    df['ProductSize'] = sizes
    return df

# Run the async function and get result
import nest_asyncio
nest_asyncio.apply()
df_result = await fetch_all_sizes(df)
df_result
```
