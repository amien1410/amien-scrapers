```python
import requests
from bs4 import BeautifulSoup

url = "https://www.warrensaunders.com.au"
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())  # Or: print(soup.get_text())
```

```python

from urllib.parse import urljoin

# Extract all links
links = set()
for a_tag in soup.find_all("a", href=True):
    full_url = urljoin(url, a_tag["href"])
    links.add(full_url)

# Print unique links
for link in sorted(links):
    print(link)
```

```python
# Find all divs with class 'content-block'
content_blocks = soup.find_all("div", class_="content-block")

# Print the text from each block
for i, block in enumerate(content_blocks, start=1):
    print(f"\n--- Block {i} ---\n")
    print(block.get_text(strip=True, separator="\n"))
```

```python

# Extract text from each block
text_output = []
for block in content_blocks:
    text = block.get_text(separator="\n", strip=True)
    if text:
        text_output.append(text)

# Save to a text file
with open("warrensaunders_content.txt", "w", encoding="utf-8") as file:
    file.write("\n\n---\n\n".join(text_output))

print("Content saved to warrensaunders_content.txt")
```
