```python
"""
Timeline.json -> Google Maps Visualizer + Street-by-Street Log
================================================================
Designed to run in Google Colab.

What it does:
1. Loads a Google Location History "timeline.json" export
   (the semanticSegments / rawSignals format shown in your sample).
2. Walks every point chronologically (timelinePath points, activity
   start/end points, and parking points).
3. Reverse-geocodes each point using the Google Geocoding API to get
   a real street address (cached so repeat coordinates are cheap).
4. Prints + saves a GPS-style "trip log": timestamp, street address,
   and activity info (mode, distance, parking, etc).
5. Renders an interactive Google Map (Maps JavaScript API) inline in
   Colab with numbered markers, info windows (time + address) and a
   polyline connecting the route in order.

Setup
-----
1. Get an API key at https://console.cloud.google.com/apis/credentials
   Enable these APIs on the project: "Maps JavaScript API" and
   "Geocoding API".
2. Upload your timeline.json to the Colab session (left sidebar ->
   Files -> upload), or mount Drive and point TIMELINE_PATH at it.
3. Paste your key into API_KEY below.
4. Run all cells.

Note: this script only reads your own location export -- nothing is
sent anywhere except Google's own Maps/Geocoding APIs using your key.
"""

# ============================================================
# CELL 1 - Install dependencies (Colab already has most of these)
# ============================================================
!pip install -q googlemaps

# ============================================================
# CELL 2 - Imports & config
# ============================================================
import json
import time
from datetime import datetime
from dateutil import parser as dateparser
import googlemaps
from IPython.display import HTML, display

API_KEY = "YOUR_GOOGLE_MAPS_API_KEY"          # <-- put your key here
TIMELINE_PATH = "timeline.json"                # <-- path to uploaded file
OUTPUT_LOG_PATH = "trip_log.txt"

gmaps_client = googlemaps.Client(key=API_KEY)

# ============================================================
# CELL 3 - Load and parse timeline.json
# ============================================================
def load_timeline(path):
    with open(path, "r") as f:
        return json.load(f)

def parse_latlng(point_str):
    """'40.9138182°, -74.0638793°' -> (40.9138182, -74.0638793)"""
    lat_str, lng_str = point_str.replace("°", "").split(",")
    return float(lat_str.strip()), float(lng_str.strip())

def build_point_list(data):
    """
    Returns a chronologically sorted list of dicts:
    {time, lat, lng, kind, extra}
    kind is one of: 'path', 'activity_start', 'activity_end', 'parking'
    """
    points = []

    for segment in data.get("semanticSegments", []):
        # Raw GPS breadcrumb points (walking/driving path)
        for p in segment.get("timelinePath", []):
            lat, lng = parse_latlng(p["point"])
            points.append({
                "time": dateparser.isoparse(p["time"]),
                "lat": lat,
                "lng": lng,
                "kind": "path",
                "extra": None,
            })

        # Activity segments (trip start/end + mode + parking)
        activity = segment.get("activity")
        if activity:
            mode = activity.get("topCandidate", {}).get("type", "UNKNOWN")
            distance = activity.get("distanceMeters")

            if "start" in activity:
                lat, lng = parse_latlng(activity["start"]["latLng"])
                points.append({
                    "time": dateparser.isoparse(segment["startTime"]),
                    "lat": lat,
                    "lng": lng,
                    "kind": "activity_start",
                    "extra": f"Trip start ({mode}, {distance:.0f} m)" if distance else f"Trip start ({mode})",
                })

            if "end" in activity:
                lat, lng = parse_latlng(activity["end"]["latLng"])
                points.append({
                    "time": dateparser.isoparse(segment["endTime"]),
                    "lat": lat,
                    "lng": lng,
                    "kind": "activity_end",
                    "extra": f"Trip end ({mode}, {distance:.0f} m)" if distance else f"Trip end ({mode})",
                })

            parking = activity.get("parking")
            if parking:
                lat, lng = parse_latlng(parking["location"]["latLng"])
                points.append({
                    "time": dateparser.isoparse(parking["startTime"]),
                    "lat": lat,
                    "lng": lng,
                    "kind": "parking",
                    "extra": "Parked here",
                })

    points.sort(key=lambda x: x["time"])
    return points


# ============================================================
# CELL 4 - Reverse geocode each point (cached)
# ============================================================
def reverse_geocode_points(points, gclient, sleep_between_calls=0.05):
    cache = {}
    for pt in points:
        key = (round(pt["lat"], 6), round(pt["lng"], 6))
        if key in cache:
            pt["address"] = cache[key]
            continue
        try:
            results = gclient.reverse_geocode((pt["lat"], pt["lng"]))
            address = results[0]["formatted_address"] if results else "Unknown location"
        except Exception as e:
            address = f"(geocode failed: {e})"
        cache[key] = address
        pt["address"] = address
        time.sleep(sleep_between_calls)  # be gentle on rate limits
    return points


# ============================================================
# CELL 5 - Build the written, GPS-style log
# ============================================================
KIND_LABELS = {
    "path": "En route",
    "activity_start": "Trip start",
    "activity_end": "Trip end",
    "parking": "Parked",
}

def build_log_text(points):
    lines = []
    lines.append("=" * 70)
    lines.append("TRIP LOG")
    lines.append("=" * 70)
    for i, pt in enumerate(points, 1):
        ts = pt["time"].strftime("%Y-%m-%d %I:%M:%S %p %Z")
        label = KIND_LABELS.get(pt["kind"], pt["kind"])
        line = f"[{i:03d}] {ts}  |  {label:12s}  |  {pt['address']}"
        if pt["extra"]:
            line += f"  ({pt['extra']})"
        lines.append(line)
    return "\n".join(lines)

def save_and_print_log(points, out_path):
    text = build_log_text(points)
    print(text)
    with open(out_path, "w") as f:
        f.write(text)
    print(f"\nSaved log to {out_path}")


# ============================================================
# CELL 6 - Render an interactive Google Map inline in Colab
# ============================================================
def build_map_html(points, api_key, zoom=15):
    if not points:
        return "<p>No points to show.</p>"

    center_lat = points[len(points) // 2]["lat"]
    center_lng = points[len(points) // 2]["lng"]

    # Build JS array of markers
    marker_js = []
    path_js = []
    for i, pt in enumerate(points, 1):
        ts = pt["time"].strftime("%Y-%m-%d %I:%M:%S %p")
        label = KIND_LABELS.get(pt["kind"], pt["kind"])
        info = f"#{i} {label}<br>{ts}<br>{pt['address']}"
        if pt["extra"]:
            info += f"<br><i>{pt['extra']}</i>"
        info_escaped = info.replace("'", "\\'")

        color = {
            "path": "#4285F4",
            "activity_start": "#0F9D58",
            "activity_end": "#DB4437",
            "parking": "#F4B400",
        }.get(pt["kind"], "#4285F4")

        marker_js.append(f"""
            new google.maps.Marker({{
                position: {{lat: {pt['lat']}, lng: {pt['lng']}}},
                map: map,
                title: '#{i} {label}',
                icon: {{
                    path: google.maps.SymbolPath.CIRCLE,
                    scale: 6,
                    fillColor: '{color}',
                    fillOpacity: 1,
                    strokeWeight: 1,
                    strokeColor: '#ffffff'
                }}
            }}).addListener('click', function() {{
                infoWindows[{i-1}].open(map, this);
            }});
        """)
        path_js.append(f"{{lat: {pt['lat']}, lng: {pt['lng']}}}")

    infowindow_js = "\n".join(
        f"new google.maps.InfoWindow({{content: '{ (KIND_LABELS.get(p['kind'], p['kind']) + ' #' + str(i)) }: {p['address'].replace(chr(39), chr(92)+chr(39))}'}})"
        for i, p in enumerate(points, 1)
    )

    html = f"""
    <div id="map" style="height: 600px; width: 100%;"></div>
    <script>
      function initMap() {{
        var map = new google.maps.Map(document.getElementById('map'), {{
          zoom: {zoom},
          center: {{lat: {center_lat}, lng: {center_lng}}},
          mapTypeId: 'roadmap'
        }});

        var infoWindows = [
          {infowindow_js}
        ];

        {' '.join(marker_js)}

        var routePath = new google.maps.Polyline({{
          path: [{', '.join(path_js)}],
          geodesic: true,
          strokeColor: '#4285F4',
          strokeOpacity: 0.8,
          strokeWeight: 3
        }});
        routePath.setMap(map);
      }}
    </script>
    <script src="https://maps.googleapis.com/maps/api/js?key={api_key}&callback=initMap&loading=async" async defer></script>
    """
    return html


# ============================================================
# CELL 7 - Run everything
# ============================================================
data = load_timeline(TIMELINE_PATH)
points = build_point_list(data)
points = reverse_geocode_points(points, gmaps_client)

save_and_print_log(points, OUTPUT_LOG_PATH)

map_html = build_map_html(points, API_KEY)
display(HTML(map_html))
```
