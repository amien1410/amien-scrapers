```python
import requests

# Retrieve your API key from Colab Secrets
CLEARBIT_API_KEY = ""

# If you prefer to hardcode for testing (NOT RECOMMENDED for production)
# CLEARBIT_API_KEY = 'sk_YOUR_API_KEY_HERE'

def clearbit(email):
    if not CLEARBIT_API_KEY:
        print("Clearbit API key not found. Please set it in Colab Secrets.")
        return

    url = f'https://person-stream.clearbit.com/v2/combined/find?email={email}'
    headers = {
        "Authorization": f"Bearer {CLEARBIT_API_KEY}"
    }

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status() # Raise an exception for HTTP errors (4xx or 5xx)
        data = response.json()

        print("API Response:")
        print(data)

        person = data.get('person')
        company = data.get('company')

        info = ""

        if person:
            bio = person.get('bio', '')
            location = person.get('location', '')
            info += f"{bio} {location} ".strip()
            print(f"Person Bio and Location: {bio} {location}")

            employment = person.get('employment')
            if employment:
                title = employment.get('title', '')
                info += f"{title} ".strip()
                print(f"Person Title: {title}")

            linkedin = person.get('linkedin')
            if linkedin:
                handle = linkedin.get('handle', '')
                info += f"{handle} ".strip()
                print(f"Person LinkedIn: https://linkedin.com/{handle}")

        if company:
            domain = company.get('domain', '')
            info += f"{domain} ".strip()
            print(f"Company Domain: {domain}")

        print(f"\nAggregated Info: {info.strip()}")

    except requests.exceptions.HTTPError as errh:
        print(f"HTTP Error: {errh}")
    except requests.exceptions.ConnectionError as errc:
        print(f"Error Connecting: {errc}")
    except requests.exceptions.Timeout as errt:
        print(f"Timeout Error: {errt}")
    except requests.exceptions.RequestException as err:
        print(f"An unknown error occurred: {err}")
    except Exception as e:
        print(f"An error occurred while processing the response: {e}")
```

Now you can call the function with an email address:

```python
# Example usage (replace with a real email for Clearbit lookup)
clearbit('alex@alexmaccaw.com')
```
