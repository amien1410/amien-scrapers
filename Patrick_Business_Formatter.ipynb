### Combine all CSV files into a single DataFrame

```python
import pandas as pd
import os

# Define the directory where CSV files are located
csv_directory = '/content/drive/MyDrive/Upwork/patrick-bwn-project/Google Places API - Part 3'

# List all files in the directory and filter for .csv files
csv_files = [f for f in os.listdir(csv_directory) if f.endswith('.csv')]
```

```python
# Read each CSV file into a DataFrame and store them in a list
df_list = []
for filename in csv_files:
    file_path = os.path.join(csv_directory, filename)
    try:
        df = pd.read_csv(file_path)
        df_list.append(df)
        print(f"Successfully loaded {filename}")
    except Exception as e:
        print(f"Error loading {filename}: {e}")

# Concatenate all DataFrames in the list into a single DataFrame
if df_list:
    combined_df = pd.concat(df_list, ignore_index=True)
    print("\nAll CSV files combined successfully!")
else:
    combined_df = pd.DataFrame()
    print("No CSV files were found or loaded.")
```

### Saving the combined data to a CSV file

```python
# Save the combined DataFrame to a new CSV file
output_csv_filename = 'final_results_part3.csv'
combined_df.to_csv(output_csv_filename, index=False)

print(f"Combined data saved to {output_csv_filename}")
```

### Exploring Unique Values in Each Column

```python
# Iterate through each column to find unique values and their counts
for column in combined_df.columns:
    unique_values_count = combined_df[column].nunique()
    print(f"\nColumn: {column}")
    print(f"Number of unique values: {unique_values_count}")
    if unique_values_count < 20:  # Display all unique values if there are less than 20
        print("Unique values and their counts:")
        print(combined_df[column].value_counts(dropna=False))
    else: # If there are too many unique values, display the top 10
        print("Top 10 unique values and their counts:")
        print(combined_df[column].value_counts(dropna=False).head(10))
```

### Combine Multiple CSV Files and Remove Duplicates

```python
import pandas as pd
import os

# Define the paths to the input CSV files
file_paths = [
    "/content/drive/MyDrive/Upwork/patrick-bwn-project/Google Places API/final_result.csv",
    "/content/drive/MyDrive/Upwork/patrick-bwn-project/Google Places API - Part 2/final_results_part2.csv",
    "/content/drive/MyDrive/Upwork/patrick-bwn-project/Google Places API - Part 3/final_results_part3.csv"
]

# List to hold individual DataFrames
df_list = []

# Read each CSV file into a DataFrame and add to the list
for file_path in file_paths:
    try:
        # Added low_memory=False to handle DtypeWarning if columns have mixed types
        df = pd.read_csv(file_path, low_memory=False)
        df_list.append(df)
        print(f"Successfully loaded: {file_path}")
    except Exception as e:
        print(f"Error loading {file_path}: {e}")

# Concatenate all DataFrames into a single DataFrame
if df_list:
    combined_all_df = pd.concat(df_list, ignore_index=True)
    print(f"\nSuccessfully combined {len(df_list)} CSV files.")
else:
    combined_all_df = pd.DataFrame()
    print("No DataFrames were loaded.")

# Display the initial shape of the combined DataFrame
print(f"Initial combined DataFrame shape: {combined_all_df.shape}")
```

```python
# Remove duplicates based on the 'place_id' column
initial_rows = combined_all_df.shape[0]
combined_all_df.drop_duplicates(subset=['place_id'], inplace=True)
final_rows = combined_all_df.shape[0]

print(f"\nRemoved {initial_rows - final_rows} duplicate rows based on 'place_id'.")
print(f"Final DataFrame shape after removing duplicates: {combined_all_df.shape}")

# Display the first few rows of the final DataFrame
display(combined_all_df.head())
```

```python
# Save the final DataFrame to a new CSV file
output_filename = "all_results.csv"
combined_all_df.to_csv(output_filename, index=False)

print(f"\nFinal combined and deduplicated data saved to '{output_filename}'.")
```

### Combining TripAdvisor CSV Files and Removing Duplicates

```python
import pandas as pd
import os

# Define the directories where CSV files are located
directories = [
    "/content/drive/MyDrive/Upwork/patrick-bwn-project/TripAdvisor - State Based Search"
]

# List to hold individual DataFrames
df_list = []

# Iterate through each directory
for directory in directories:
    if not os.path.isdir(directory):
        print(f"Warning: Directory not found: {directory}")
        continue

    # List all files in the directory and filter for '_all_categories.csv' files
    csv_files_in_dir = [f for f in os.listdir(directory) if f.endswith('_all_categories.csv')]

    print(f"\nProcessing directory: {directory}")
    if not csv_files_in_dir:
        print("No matching CSV files found in this directory.")
        continue

    # Read each matching CSV file into a DataFrame and add to the list
    for filename in csv_files_in_dir:
        file_path = os.path.join(directory, filename)
        try:
            df = pd.read_csv(file_path, low_memory=False)
            df_list.append(df)
            print(f"Successfully loaded: {filename}")
        except Exception as e:
            print(f"Error loading {file_path}: {e}")

# Concatenate all DataFrames into a single DataFrame
if df_list:
    combined_tripadvisor_df = pd.concat(df_list, ignore_index=True)
    print(f"\nSuccessfully combined {len(df_list)} TripAdvisor CSV files.")
else:
    combined_tripadvisor_df = pd.DataFrame()
    print("No TripAdvisor DataFrames were loaded.")

# Display the initial shape of the combined DataFrame
print(f"Initial combined TripAdvisor DataFrame shape: {combined_tripadvisor_df.shape}")
```

```python
# Remove duplicates based on the 'detail_id' column
if not combined_tripadvisor_df.empty and 'detail_id' in combined_tripadvisor_df.columns:
    initial_rows = combined_tripadvisor_df.shape[0]
    combined_tripadvisor_df.drop_duplicates(subset=['detail_id'], inplace=True)
    final_rows = combined_tripadvisor_df.shape[0]

    print(f"\nRemoved {initial_rows - final_rows} duplicate rows based on 'detail_id'.")
    print(f"Final DataFrame shape after removing duplicates: {combined_tripadvisor_df.shape}")
else:
    print("Cannot remove duplicates: DataFrame is empty or 'detail_id' column is missing.")

# Display the first few rows of the final DataFrame
display(combined_tripadvisor_df.head())
```

```python
# Save the final DataFrame to a new CSV file
output_filename = "tripadvisor_all_state_results.csv"
if not combined_tripadvisor_df.empty:
    combined_tripadvisor_df.to_csv(output_filename, index=False)
    print(f"\nFinal combined and deduplicated TripAdvisor data saved to '{output_filename}'.")
else:
    print("No data to save: combined_tripadvisor_df is empty.")
```

```python
import pandas as pd
import os

# Define the directory where JSON files are located
json_directory = "/content/drive/MyDrive/Upwork/patrick-bwn-project/TripAdvisor - State Based Search"

# List all files in the directory and filter for '_summary_report.json' files
json_files = [f for f in os.listdir(json_directory) if f.endswith('_summary_report.json')]

# List to hold individual DataFrames
df_json_list = []

print(f"Found {len(json_files)} summary report JSON files in {json_directory}")

# Read each JSON file into a DataFrame and store them in a list
for filename in json_files:
    file_path = os.path.join(json_directory, filename)
    try:
        # Read JSON and treat categories as index, states as columns
        df_json = pd.read_json(file_path)
        df_json_list.append(df_json)
        print(f"Successfully loaded {filename}")
    except Exception as e:
        print(f"Error loading {filename}: {e}")

# Concatenate all DataFrames along columns, performing an outer join on the index
if df_json_list:
    combined_summary_df = pd.concat(df_json_list, axis=1, join='outer')
    print("\nAll summary report JSON files combined successfully!")
else:
    combined_summary_df = pd.DataFrame()
    print("No summary report JSON files were found or loaded.")

# Display the first 5 rows of the combined DataFrame
display(combined_summary_df.head())
```
