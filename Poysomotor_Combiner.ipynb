```python
import os

# Root directory where your folders are located
root_dir = r"/content/root"
output_file = "combined_result2.txt"

# Dictionary to store contents grouped by folder name
folder_contents = {}

# Walk through all directories and files
for dirpath, dirnames, filenames in os.walk(root_dir):
    for file in filenames:
        if file.lower().endswith(".txt"):
            folder_name = os.path.basename(dirpath)  # Get the folder name
            file_path = os.path.join(dirpath, file)

            with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
                content = f.read()

            folder_contents.setdefault(folder_name, []).append((file, content))

# Write combined file
with open(output_file, "w", encoding="utf-8") as out:
    for folder, files in folder_contents.items():
        out.write(f"\n{'='*50}\n")
        out.write(f"SECTION: {folder}\n")
        out.write(f"{'='*50}\n\n")
        for filename, text in files:
            out.write(f"===== START FILE: {filename} =====\n")
            out.write(text.strip() + "\n")
            out.write(f"===== END FILE: {filename} =====\n\n")

print(f"✅ Combined file created: {output_file}")
```
