```python
#@title Facebook - Rapidapi (facebook-pages-scraper2)
import requests

url = "https://facebook-pages-scraper2.p.rapidapi.com/get_facebook_marketplace_items_listing"
querystring = {"query":"cars","filter_location_latitude":"40.7142","filter_location_longitude":"-74.0064","filter_price_lower_bound":"0","filter_price_upper_bound":"214748364700","filter_radius_km":"65"}
headers = {
    "x-rapidapi-host": "facebook-pages-scraper2.p.rapidapi.com",
    "x-rapidapi-key": ""
}

response = requests.get(url, headers=headers, params=querystring)
response.json()
```
