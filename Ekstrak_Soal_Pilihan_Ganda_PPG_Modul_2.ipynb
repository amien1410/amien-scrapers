```python
soup = BeautifulSoup(html, "html.parser")

# find all soal blocks
soal_divs = soup.find_all("div", id=lambda x: x and x.startswith("soal-"))

results = []

for div in soal_divs:
    soal_data = {}

    # nomor soal
    nomor = div.find("h4", class_="text-headline").get_text(strip=True)
    nomor = nomor.replace("Soal Nomor", "").strip()
    soal_data["nomor"] = nomor

    # teks soal
    soal_text = div.find("p", class_="text-body-2").get_text(strip=True)
    soal_data["soal"] = soal_text

    # pilihan (A-E)
    pilihan_labels = div.find_all("label")
    for lbl in pilihan_labels:
        text = lbl.get_text(strip=True)
        if text.startswith("A."):
            soal_data["A"] = text[2:].strip()
        elif text.startswith("B."):
            soal_data["B"] = text[2:].strip()
        elif text.startswith("C."):
            soal_data["C"] = text[2:].strip()
        elif text.startswith("D."):
            soal_data["D"] = text[2:].strip()
        elif text.startswith("E."):
            soal_data["E"] = text[2:].strip()

    results.append(soal_data)

# tampilkan hasil
for r in results:
    print(r)
```
