```python
from requests.auth import HTTPBasicAuth
import requests

client_id = ''
client_secret = ''

token_url = "https://api.ebay.com/identity/v1/oauth2/token"
headers_token = {"Content-Type": "application/x-www-form-urlencoded"}
body = {
    "grant_type": "client_credentials",
    "scope": "https://api.ebay.com/oauth/api_scope"
}
response_token = requests.post(token_url, headers=headers_token, data=body,
                                auth=HTTPBasicAuth(client_id, client_secret))

if response_token.status_code != 200:
    print("error getting token")

token = response_token.json()['access_token']

headers = {
    'Authorization': f'Bearer {token}',
    'X-EBAY-C-ENDUSERCTX': 'contextualLocation=country=IE',
    'X-EBAY-C-MARKETPLACE-ID': 'EBAY_GB',
    'Content-Type': 'application/json'
}

url = f"https://api.ebay.com/buy/browse/v1/item_summary/search?q=ps5&limit=20"
response = requests.get(url, headers=headers)

if response.status_code != 200:
    print("error: token invalid")

data = response.json()
data
```
