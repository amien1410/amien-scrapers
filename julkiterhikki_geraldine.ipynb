```python
import requests
import json
import time

# ---------------- CONFIG ----------------
TARGET_RECORDS = 5000
SAVE_EVERY = 100
PREFIXES = ["0310291", "0320251"]
OUTPUT_FILE = "results.json"

# ---------------- SESSION ----------------
session = requests.Session()

headers = {
    "Accept-Encoding": "gzip, deflate, br, zstd",
    "Accept-Language": "en-US,en;q=0.9,id;q=0.8,fa;q=0.7,ar;q=0.6,ms;q=0.5,ja;q=0.4,es;q=0.3",
    "Content-Type": "application/transit+json",
    "Priority": "u=1, i",
    "Referer": "https://julkiterhikki.lvv.fi/",
    "Sec-Ch-Ua": "\"Microsoft Edge\";v=\"147\", \"Not.A/Brand\";v=\"8\", \"Chromium\";v=\"147\"",
    "Sec-Ch-Ua-Mobile": "?0",
    "Sec-Ch-Ua-Platform": "\"Windows\"",
    "Sec-Fetch-Dest": "empty",
    "Sec-Fetch-Mode": "cors",
    "Sec-Fetch-Site": "same-origin",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/147.0.0.0 Safari/537.36 Edg/147.0.0.0",
    "Accept": "application/transit+json" # Added as it's a common header and was present in a successful request
}

# ---------------- AUTH PAYLOADS ----------------
auth_url = "https://julkiterhikki.lvv.fi/api/authenticate"

auth_payloads = [
    ["^ ","~:id",["~:en","~:qa",0],"~:answer","dog"],
    ["^ ","~:id",["~:en","~:addition",["^ ","~:type","~:numeric","~:values",["six","six"]]],"~:answer","12"],
    ["^ ","~:id",["~:en","~:addition",["^ ","~:type","~:text","~:values",["nine","two"]]],"~:answer","eleven"],
    ["^ ","~:id",["~:en","~:addition",["^ ","~:type","~:numeric","~:values",[1,2]]],"~:answer","3"],
    ["^ ","~:id",["~:en","~:addition",["^ ","~:type","~:numeric","~:values",[1,4]]],"~:answer","5"],
    ["^ ","~:id",["~:en","~:addition",["^ ","~:type","~:numeric","~:values",[1,6]]],"~:answer","6"],
    ["^ ","~:id",["~:en","~:addition",["^ ","~:type","~:numeric","~:values",[2,2]]],"~:answer","4"],
    ["^ ","~:id",["~:en","~:qa",6],"~:answer","Saturday"],
    ["^ ","~:id",["~:en","~:qa",1],"~:answer","right"],
    ["^ ","~:id",["~:en","~:qa",3],"~:answer","Turkey"],
    ["^ ","~:id",["~:en","~:qa",7],"~:answer","April"],
    ["^ ","~:id",["~:en","~:qa",8],"~:answer","Twelve"]
]

def authenticate():
    print("🔐 Authenticating...")

    for idx, payload in enumerate(auth_payloads, 1):
        try:
            res = session.post(auth_url, data=json.dumps(payload), headers=headers)

            if res.status_code == 200:
                print(f"✅ Auth success using payload #{idx}")
                return True
            else:
                print(f"⚠️ Payload #{idx} failed (status {res.status_code})")

        except Exception as e:
            print(f"⚠️ Payload #{idx} error: {e}")

    return False


if not authenticate():
    print("❌ All authentication methods failed")
    exit()


# ---------------- PARSER ----------------
def parse_person_record(data):
    result = {}
    items = data[0]

    i = 0
    while i < len(items):
        key = items[i]

        if key == "~:etunimet":
            result["first_names"] = items[i + 1]
            i += 2

        elif key == "~:sukunimi":
            result["last_name"] = items[i + 1]
            i += 2

        elif key == "~:rekisterointinumero":
            result["registration_number"] = items[i + 1]
            i += 2

        elif key == "~:syntynyt":
            result["birth_year"] = items[i + 1]
            i += 2

        elif key == "~:oikeudet":
            rights_list = items[i + 1]
            parsed_rights = []

            for r in rights_list:
                right_dict = {}
                j = 0
                while j < len(r):
                    if r[j] == "~:nimi":
                        name_block = r[j + 1]
                        k = 0
                        while k < len(name_block):
                            if name_block[k] == "~:EN":
                                right_dict["title_en"] = name_block[k + 1]
                            k += 1
                        j += 2

                    elif r[j] == "~:rajoituksia":
                        right_dict["has_restrictions"] = r[j + 1]
                        j += 2

                    else:
                        j += 1

                parsed_rights.append(right_dict)

            result["rights"] = parsed_rights
            i += 2

        else:
            i += 1

    return result


# ---------------- SEARCH LOOP ----------------
search_url = "https://julkiterhikki.lvv.fi/api/search"

results = []
count = 0

for prefix in PREFIXES:
    number = 0

    while count < TARGET_RECORDS:
        reg_num = f"{prefix}{str(number).zfill(4)}"
        print(f"Checking reg number: {reg_num}")
        payload = ["^ ", "~:rekisterointinumero", reg_num]

        try:
            res = session.post(search_url, data=json.dumps(payload), headers=headers)

            if res.status_code != 200:
                number += 1
                continue

            try:
                data = res.json()
            except:
                number += 1
                continue

            if not data or not isinstance(data, list):
                number += 1
                continue

            print(data)
            parsed = parse_person_record(data)

            if "registration_number" not in parsed:
                number += 1
                continue

            results.append(parsed)
            count += 1

            print(f"✅ {count} | {reg_num}")

            # Save every 100
            if count % SAVE_EVERY == 0:
                with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
                    json.dump(results, f, indent=2, ensure_ascii=False)
                print(f"💾 Saved {count} records")

            if count >= TARGET_RECORDS:
                break

            time.sleep(0.2)

        except Exception as e:
            print(f"⚠️ Error on {reg_num}: {e}")

        number += 1


# ---------------- FINAL SAVE ----------------
with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    json.dump(results, f, indent=2, ensure_ascii=False)

print(f"\n🎉 DONE: {count} records saved")
```
