```python
import os
import json

# Directory where the JSON files are located
json_dir = '/content/'

# Dictionary to store the loaded JSON data
loaded_json_data = {}

# Iterate through all files in the directory
for filename in os.listdir(json_dir):
    if filename.endswith('.json'):
        filepath = os.path.join(json_dir, filename)
        # Extract variable name from filename (remove .json extension)
        var_name = os.path.splitext(filename)[0]
        try:
            with open(filepath, 'r', encoding='utf-8') as f:
                loaded_json_data[var_name] = json.load(f)
            print(f"Successfully loaded '{filename}' into variable '{var_name}'.")
        except json.JSONDecodeError as e:
            print(f"Error loading '{filename}': The file is not a valid JSON. Please check its content. Error details: {e}")
        except Exception as e:
            print(f"An unexpected error occurred while loading '{filename}': {e}")

# Display the keys of the dictionary to show what data has been loaded
print("\nVariables containing loaded JSON data:")
for key in loaded_json_data.keys():
    print(f"- {key}")
```

The `loaded_json_data` dictionary now contains the content of all JSON files found in the `/content/` folder. You can access them using their filenames (without the `.json` extension) as keys. For example, to view the content of `agedcare.json`, you would use `loaded_json_data['agedcare']`.

```python
def safe_text(obj):
    return obj.get('text') if isinstance(obj, dict) else None

def extract_linkedin_companies(data):
    companies = []

    for item in data:
        if item.get('$type') != 'com.linkedin.voyager.dash.search.EntityResultViewModel':
            continue

        title_obj = item.get('title')
        subtitle_obj = item.get('primarySubtitle')
        secondary_obj = item.get('secondarySubtitle')
        summary_obj = item.get('summary')

        subtitle_text = safe_text(subtitle_obj) or ""
        industry = ""
        location = ""

        if "•" in subtitle_text:
            parts = subtitle_text.split("•", 1)
            industry = parts[0].strip()
            location = parts[1].strip()
        else:
            industry = subtitle_text

        company_info = {
            'name': safe_text(title_obj),
            'industry': industry,
            'location': location,
            'followers': safe_text(secondary_obj),
            'linkedin_url': item.get('navigationUrl'),
            'summary': safe_text(summary_obj),
            'tracking_urn': item.get('trackingUrn'),
            'logo_url': extract_logo_url(item)
        }

        companies.append(company_info)

    return companies

def extract_logo_url(item):
    attributes = item.get('image', {}).get('attributes', [])
    if not attributes:
        return None

    detail_data = attributes[0].get('detailData', {})
    if not detail_data:
        return None

    logo_data = detail_data.get('nonEntityCompanyLogo')
    if not logo_data:
        return None

    vector_img = logo_data.get('vectorImage')
    if not vector_img:
        return None

    artifacts = vector_img.get('artifacts')
    if not artifacts:
        return None

    return artifacts[0].get('fileIdentifyingUrlPathSegment')


# Usage:
results = extract_linkedin_companies(loaded_json_data['medical'][0]['included'])
for c in results: print(c)
```

```python
all_results = {}

for category, pages in loaded_json_data.items():
    print(f"\n=== Processing category: {category} ===")

    category_results = []

    for page_index, page_data in enumerate(pages, start=1):
        included = page_data.get("included", [])
        if not included:
            continue

        results = extract_linkedin_companies(included)
        category_results.extend(results)

        print(f"  Page {page_index}: {len(results)} companies")

    all_results[category] = category_results

print("\n✅ Processing complete")
```

```python
import pandas as pd

# Initialize an empty list to store DataFrames
df_list = []

# Iterate through each category in all_results
for category, companies_data in all_results.items():
    # Create a DataFrame for the current category
    if companies_data:
        df_category = pd.DataFrame(companies_data)
        df_category['category'] = category  # Add a category column
        df_list.append(df_category)

# Concatenate all DataFrames into a single DataFrame
if df_list:
    all_companies_df = pd.concat(df_list, ignore_index=True)
    print("Combined DataFrame created successfully!")
    print("First 5 rows of the combined DataFrame:")
    display(all_companies_df.head())
    print(f"Total companies in DataFrame: {len(all_companies_df)}")
else:
    print("No data available to create a DataFrame.")
    all_companies_df = pd.DataFrame()
```

```python
# Save the DataFrame to a CSV file
csv_filename = 'all_companies.csv'
if not all_companies_df.empty:
    all_companies_df.to_csv(csv_filename, index=False, encoding='utf-8')
    print(f"DataFrame successfully saved to {csv_filename}")
else:
    print("DataFrame is empty, nothing to save.")
```
