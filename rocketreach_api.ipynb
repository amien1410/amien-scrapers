```python
import requests
import pprint

url = "https://api.rocketreach.co/v2/api/search"

headers = {
    "accept": "application/json",
    "Api-Key": "",
    "Content-Type": "application/json"
}

payload = {
    "query": {
        "name": ["\"Gerard Foley\""],          # exact match
        "current_employer": ["\"IPN Medical Centres\""],  # exact employer only
        "location": ["\"Sydney\"::~50mi"]          # exact country
    },
    "start": 1,
    "page_size": 10,
    "order_by": "relevance"
}

response = requests.post(url, json=payload, headers=headers)

pprint.pprint(response.json())
```
