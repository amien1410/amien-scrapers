```python
import aiohttp
import asyncio

async def fetch_boats_aiohttp(page=1):
    """
    Fetch one page of boat listings asynchronously using aiohttp.
    """
    params_copy = DEFAULT_PARAMS.copy()
    params_copy["page"] = page

    # Use the same headers as the last attempt with requests
    async with aiohttp.ClientSession(headers=HEADERS) as session:
        async with session.get(BASE_URL, params=params_copy, timeout=60) as response:
            response.raise_for_status() # Raise an exception for HTTP errors
            return await response.json()

# Example usage for the new asynchronous function
async def main():
    try:
        data = await fetch_boats_aiohttp(page=1)

        # Print summary
        print("Total Listings:", data.get("count"))
        print("Listings Returned:", len(data.get("records", [])))

        # Show first 5 boats
        for boat in data.get("records", [])[:5]:
            year = boat.get("year")
            make = boat.get("make")
            model = boat.get("model")
            price = boat.get("price", {}).get("type", {}).get("amount")

            print(f"{year} {make} {model} - ${price}")

        # Print the generated URL for debugging
        params_copy = DEFAULT_PARAMS.copy()
        params_copy["page"] = 1
        print("\nGenerated URL (for aiohttp call):")
        print(f"{BASE_URL}?{urlencode(params_copy)}")

    except aiohttp.ClientResponseError as e:
        print(f"HTTP Error: {e.status} {e.message} for URL: {e.request_info.url}")
    except aiohttp.ClientConnectorError as e:
        print(f"Connection Error: {e}")
    except asyncio.TimeoutError:
        print("Request timed out.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

await main()
```
