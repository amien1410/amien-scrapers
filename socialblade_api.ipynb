```python
import requests

url = "https://matrix.sbapis.com/b/instagram/top"

headers = {
    "clientid": "",
    "token": ""
}

params = {
    "query": "followers",
    "page": 0
}

response = requests.get(url, headers=headers, params=params)

if response.ok:
    data = response.json()
    print(data)
else:
    print(f"Error {response.status_code}: {response.text}")
```
