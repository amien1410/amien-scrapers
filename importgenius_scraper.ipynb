```python
#@title Import Genius Sitemap
url = "https://www.importgenius.com/sitemap.xml"
res = requests.get(url, headers=headers)
# Parse the XML data from res.text
root = ET.fromstring(res.text)

# Find all 'loc' elements and extract their text
urls = []
for url_element in root.findall('{http://www.sitemaps.org/schemas/sitemap/0.9}url'):
    loc_element = url_element.find('{http://www.sitemaps.org/schemas/sitemap/0.9}loc')
    if loc_element is not None:
        print(f"Url: {loc_element.text}")
        urls.append(loc_element.text)

# Save the extracted URLs to a text file
filename = "importgenius-sitemap-urls"
with open(f'{filename}.txt', 'w') as f:
    for url in urls:
        f.write(url + '\n')

print(f"{len(urls)} URLs saved to {filename}.txt")
```

```python
#@title Scrape Importer
from bs4 import BeautifulSoup

url = "https://www.importgenius.com/importers/sinomet-sprl"

res = requests.get(url, headers=headers)
soup = BeautifulSoup(res.text, 'html.parser')
soup
```

```python
# 1 Importer basic info
company_name = soup.select_one(".company-name").get_text(strip=True)
company_desc = soup.select_one(".company-desc").get_text(strip=True) if soup.select_one(".company-desc") else None
company_address = soup.select_one(".company-address").get_text(" ", strip=True) if soup.select_one(".company-address") else None

# 2 Summary stats
def get_summary_value(label):
    el = soup.find("div", class_="company-summary-label", string=lambda x: x and label.lower() in x.lower())
    if el:
        val = el.find_next("div", class_="company-summary-value")
        return val.get_text(strip=True) if val else None
    return None

total_shipments = get_summary_value("Total Shipments")
trading_partners = get_summary_value("Trading Partners")
top_partner = get_summary_value("Top Partner")
ports = get_summary_value("Ports")
top_port = get_summary_value("Top Port")

# 3 Table of individual shipments
shipments = []
table = soup.find("table", class_="table text-nowrap")
if table:
    rows = table.select("tbody tr")
    for row in rows:
        shipment = {
            "Bill of Lading": row.find("td", {"data-title": "Bill of Lading"}).get_text(strip=True) if row.find("td", {"data-title": "Bill of Lading"}) else None,
            "Product": row.find("td", {"data-title": "Product"}).get_text(strip=True) if row.find("td", {"data-title": "Product"}) else None,
            "Importer": row.find("td", {"data-title": "Importer"}) or row.find("td", {"data-title": "Consignee"}),
            "Supplier": row.find("td", {"data-title": "Shipper"}).get_text(strip=True) if row.find("td", {"data-title": "Shipper"}) else None,
            "Arrival Date": row.find("td", {"data-title": "Arrival Date"}).get_text(strip=True) if row.find("td", {"data-title": "Arrival Date"}) else None,
            "Country of Origin": row.find("td", {"data-title": "Country of Origin"}).get_text(strip=True) if row.find("td", {"data-title": "Country of Origin"}) else None,
            "Gross Weight Kgs": row.find("td", {"data-title": "Gross Weight Kgs"}).get_text(strip=True) if row.find("td", {"data-title": "Gross Weight Kgs"}) else None,
            "Quantity": row.find("td", {"data-title": "Quantity"}).get_text(strip=True) if row.find("td", {"data-title": "Quantity"}) else None
        }
        if shipment["Importer"]:
            shipment["Importer"] = shipment["Importer"].get_text(strip=True)
        shipments.append(shipment)

# 4 Compile everything
importer_data = {
    "company_name": company_name,
    "description": company_desc,
    "address": company_address,
    "summary": {
        "total_shipments": total_shipments,
        "trading_partners": trading_partners,
        "top_partner": top_partner,
        "ports": ports,
        "top_port": top_port
    },
    "shipments": shipments
}

importer_data
```
