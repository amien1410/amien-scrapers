```python
#@title Import Genius Sitemap
url = "https://www.importgenius.com/sitemap.xml"
res = requests.get(url, headers=headers)
# Parse the XML data from res.text
root = ET.fromstring(res.text)

# Find all 'loc' elements and extract their text
urls = []
for url_element in root.findall('{http://www.sitemaps.org/schemas/sitemap/0.9}url'):
    loc_element = url_element.find('{http://www.sitemaps.org/schemas/sitemap/0.9}loc')
    if loc_element is not None:
        print(f"Url: {loc_element.text}")
        urls.append(loc_element.text)

# Save the extracted URLs to a text file
filename = "importgenius-sitemap-urls"
with open(f'{filename}.txt', 'w') as f:
    for url in urls:
        f.write(url + '\n')

print(f"{len(urls)} URLs saved to {filename}.txt")
```

```python
import re
import json

def split_top_level_commas(s):
    """Split a string at top-level commas (ignore commas inside quotes/brackets)."""
    parts = []
    cur = []
    depth = 0
    in_str = False
    esc = False
    quote_char = None
    for ch in s:
        if esc:
            cur.append(ch); esc = False; continue
        if ch == "\\":
            cur.append(ch); esc = True; continue
        if in_str:
            cur.append(ch)
            if ch == quote_char:
                in_str = False
                quote_char = None
            continue
        if ch in ('"', "'"):
            in_str = True; quote_char = ch; cur.append(ch); continue
        if ch in "([{":
            depth += 1; cur.append(ch); continue
        if ch in ")]}":
            depth -= 1; cur.append(ch); continue
        if ch == ',' and depth == 0:
            parts.append(''.join(cur).strip())
            cur = []
            continue
        cur.append(ch)
    if cur:
        parts.append(''.join(cur).strip())
    return parts

def js_to_json_compatible(s):
    """Try to make a small JS literal into JSON-compatible text for json.loads."""
    if s is None:
        return None
    t = s.strip()
    # Replace JS null/true/false to JSON literals (they are same, but be explicit)
    t = re.sub(r'\bundefined\b', 'null', t)
    t = re.sub(r'\btrue\b', 'true', t)
    t = re.sub(r'\bfalse\b', 'false', t)
    # Many args may be unquoted identifiers (like aq). We'll not convert them here.
    # Ensure double quotes for JSON strings
    if t.startswith("'") and t.endswith("'"):
        t = '"' + t[1:-1].replace('"', '\\"') + '"'
    return t

def parse_iife_and_map(script_text):
    """
    Extract the IIFE parameter list and the invocation args list
    from script_text like:
    (function(a,b,c){return {...}})(1,"x",null)
    Returns dict param->python_value (where possible), and the raw mapping.
    """
    # 1. Extract parameter list inside function(...)
    # find "function(" up to "){"
    m_params = re.search(r"\(function\s*\(\s*([^\)]*?)\s*\)\s*\{", script_text, re.DOTALL)
    if not m_params:
        raise ValueError("Could not find function parameter list.")
    params_raw = m_params.group(1).strip()
    params = [p.strip() for p in params_raw.split(',') if p.strip()]

    # 2. Extract the trailing invocation argument list: the (...) that follow the final "})"
    # We'll search for the last pattern "})(...);"
    m_args = re.search(r"\)\s*\(\s*([\s\S]*?)\s*\)\s*;?\s*$", script_text)
    if not m_args:
        # fallback: try to find the first "})(...)" (non-greedy)
        m_args = re.search(r"\)\s*\(\s*([\s\S]*?)\s*\)", script_text)
        if not m_args:
            raise ValueError("Could not find function invocation argument list.")
    args_raw = m_args.group(1).strip()
    # split top-level commas
    args_list = split_top_level_commas(args_raw)

    if len(args_list) < len(params):
        # Sometimes script passes fewer explicit values; we will still map what we have.
        pass

    mapping = {}
    for i, pname in enumerate(params):
        if i < len(args_list):
            arg_text = args_list[i]
            json_like = js_to_json_compatible(arg_text)
            parsed = None
            # try JSON loads
            try:
                parsed = json.loads(json_like)
            except Exception:
                # some args might be JS strings without quotes or numbers; try simple conversions
                t = arg_text.strip()
                # number?
                if re.fullmatch(r"-?\d+(\.\d+)?", t):
                    parsed = float(t) if '.' in t else int(t)
                # quoted string?
                elif (t.startswith('"') and t.endswith('"')) or (t.startswith("'") and t.endswith("'")):
                    parsed = t[1:-1]
                # null/undefined
                elif t in ('null', 'undefined'):
                    parsed = None
                # boolean
                elif t == 'true':
                    parsed = True
                elif t == 'false':
                    parsed = False
                else:
                    # leave raw identifier as string (e.g. "aq")
                    parsed = t
            mapping[pname] = parsed
        else:
            mapping[pname] = None

    return mapping

def replace_vars_in_payload(payload_text, mapping):
    """
    Replace bare variable tokens in payload_text with their mapped JSON values.
    Uses word-boundary replacements and json.dumps for proper representation.
    """
    out = payload_text
    # Sort by descending name length to avoid partial replacements (e.g. a and aq)
    for name in sorted(mapping.keys(), key=lambda s: -len(s)):
        val = mapping[name]
        # prepare replacement string
        if val is None:
            repl = 'null'
        elif isinstance(val, str):
            # produce a JSON string literal
            repl = json.dumps(val, ensure_ascii=False)
        else:
            repl = json.dumps(val, ensure_ascii=False)
        # replace only word-boundary occurrences (but not inside quotes)
        # We'll do a regex that ignores matches that are inside quotes by a simple heuristic:
        # (this is not perfect for all JS, but works for typical minified payloads)
        pattern = r'(?<!["\'])\b' + re.escape(name) + r'\b(?!["\'])'
        out = re.sub(pattern, repl, out)
    return out

# Step A: extract the results:{...} substring (the JS object you want to convert)
m_results = re.search(r"(results\s*:\s*\{[\s\S]*\})\s*(,|\})", target_script)
if not m_results:
    raise ValueError("Could not find results:{...} in the script text.")
results_fragment = m_results.group(1)  # text starting with 'results:{...}'

print(results_fragment)

# Step B: parse IIFE param->value mapping
mapping = parse_iife_and_map(target_script)

# Step C: replace variables inside results_fragment
replaced = replace_vars_in_payload(results_fragment, mapping)

# Step D: massage to valid JSON for parsing:
#  - quote unquoted keys
clean = re.sub(r'([,{]\s*)([A-Za-z0-9_]+\b)\s*:', r'\1"\2":', replaced)
#  - replace single quotes to double quotes (if any)
clean = clean.replace("'", '"')
#  - replace trailing commas before } or ]
clean = re.sub(r",\s*([}\]])", r"\1", clean)
#  - wrap top-level as an object if needed
if not clean.strip().startswith("{"):
    clean = "{" + clean + "}"

# Try load
try:
    data = json.loads(clean)
    print("Parsed JSON successfully.")
    # Example: what was 'aq'?
    # find mapping for 'aq'
    print("Mapping example: aq ->", mapping.get('aq'))
    # If limit was "aq" originally, after replacement it should be numeric:
    # For demonstration show the resulting limit value if present
    if "results" in data and isinstance(data["results"], dict) and "limit" in data["results"]:
        print("results.limit =", data["results"]["limit"])
    else:
        # If your fragment was 'results: {...}', you might need to access data accordingly
        print(json.dumps(data, indent=2)[:1000])
except Exception as e:
    print("Final JSON parse error:", e)
    print("Preview of cleaned text:\n", clean[:2000])
```
