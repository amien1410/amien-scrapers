## Identify Duplicate and Empty Fixed Barcodes

### Subtask:
Identify and filter all rows where the 'fixed-barcode' column has either duplicate values or is empty (NaN/null).


**Reasoning**:
To identify rows with empty or duplicate 'fixed-barcode' values, I will first create a boolean mask for empty values, then another for duplicate values (including first occurrences), combine these masks, and finally filter the DataFrame to create `df_issues`.



```python
import numpy as np

# 1. Identify rows where the 'fixed-barcode' column is null or empty
# Convert potential empty strings to NaN first to catch all empty values
df['fixed-barcode'] = df['fixed-barcode'].replace('', np.nan)
empty_barcodes = df['fixed-barcode'].isnull()

# 2. Identify rows where the 'fixed-barcode' column contains duplicate values (including the first occurrence)
duplicate_barcodes = df['fixed-barcode'].duplicated(keep=False)

# 3. Combine these two conditions to create a single boolean mask
combined_mask = empty_barcodes | duplicate_barcodes

# 4. Apply this combined mask to filter and select rows that satisfy either condition
df_issues = df[combined_mask]

# Display the first few rows of the issues DataFrame and its shape
print("Rows with empty or duplicate 'fixed-barcode' values:")
print(df_issues.head())
print(f"\nTotal rows with issues: {len(df_issues)}")
```

**Reasoning**:
The current subtask to identify rows with empty or duplicate barcodes is complete. The next logical step is to summarize these findings as requested by the main task. This includes counting the number of empty barcodes, unique duplicate barcodes, and the total number of rows with issues.



```python
empty_barcode_count = df_issues['fixed-barcode'].isnull().sum()
duplicate_barcode_count = df_issues['fixed-barcode'].duplicated(keep='first').sum()

# Count the number of unique duplicate barcode values (not including NaN)
unique_duplicate_values = df_issues['fixed-barcode'][df_issues['fixed-barcode'].duplicated(keep=False) & df_issues['fixed-barcode'].notnull()].nunique()

print("--- Summary of Fixed Barcode Issues ---")
print(f"Total rows with issues: {len(df_issues)}")
print(f"Number of rows with empty 'fixed-barcode': {empty_barcode_count}")
print(f"Number of unique duplicate 'fixed-barcode' values: {unique_duplicate_values}")
print("Note: A 'unique duplicate value' refers to a barcode string that appears more than once in the dataset.")
```

```python
import requests
import re
import json
import html
import os
import time

# -----------------------------
# CONFIG
# -----------------------------
BACKUP_EVERY = 500
PROGRESS_FILE = "progress.txt"
BACKUP_PREFIX = "backup_"
FINAL_FILE = "final_results.json"
TIMEOUT = 15
DELAY = 0.25  # be polite

# -----------------------------
# FETCH FUNCTION
# -----------------------------
def fetch_product_data(url: str) -> dict:
    headers = {
        "User-Agent": "Mozilla/5.0 (compatible; ProductScraper/1.0)",
        "Accept": "text/html",
    }

    response = requests.get(url, headers=headers, timeout=TIMEOUT)
    response.raise_for_status()

    pattern = re.compile(
        r'productData\s*:\s*(\{.*?\})\s*,\s*productCollections',
        re.DOTALL
    )

    match = pattern.search(response.text)
    if not match:
        raise ValueError("productData not found")

    raw_json = html.unescape(match.group(1))
    return json.loads(raw_json)

# -----------------------------
# LOAD PROGRESS
# -----------------------------
start_index = 0
if os.path.exists(PROGRESS_FILE):
    with open(PROGRESS_FILE, "r") as f:
        start_index = int(f.read().strip())

print(f"Starting from index: {start_index}")

# -----------------------------
# LOAD URL LIST
# -----------------------------
# urls = [...]  # YOUR 4100 URLs HERE
urls = DAOUK_URLS  # assume already defined

total = len(urls)
batch_data = []
all_files = []

# -----------------------------
# MAIN LOOP
# -----------------------------
for i in range(start_index, total):
    url = urls[i]
    print(f"Processing .... {url}")

    try:
        product = fetch_product_data(url)
        batch_data.append(product)
    except Exception:
        pass  # silent fail, resume-safe

    # Save progress
    with open(PROGRESS_FILE, "w") as f:
        f.write(str(i + 1))

    # Backup every N items
    if (i + 1) % BACKUP_EVERY == 0 or i == total - 1:
        batch_num = (i + 1) // BACKUP_EVERY + (1 if (i + 1) % BACKUP_EVERY else 0)
        backup_name = f"{BACKUP_PREFIX}{batch_num:03d}.json"

        with open(backup_name, "w", encoding="utf-8") as f:
            json.dump(batch_data, f, indent=2, ensure_ascii=False)

        all_files.append(backup_name)
        batch_data = []

        print(f"Processed {i + 1}/{total}")

    time.sleep(DELAY)

# -----------------------------
# FINAL MERGE
# -----------------------------
final_results = []

for file in sorted(f for f in os.listdir() if f.startswith(BACKUP_PREFIX)):
    with open(file, "r", encoding="utf-8") as f:
        final_results.extend(json.load(f))

with open(FINAL_FILE, "w", encoding="utf-8") as f:
    json.dump(final_results, f, indent=2, ensure_ascii=False)

print(f"\n DONE  {len(final_results)} products saved to {FINAL_FILE}")
```
