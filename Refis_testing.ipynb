```python
import redis
import urllib.parse

# The Redis connection URI provided by the user
redis_uri = ""

# Parse the URI to extract connection details
url = urllib.parse.urlparse(redis_uri)

host = url.hostname
port = url.port
username = url.username
password = url.password

try:
    # Connect to Redis using SSL and authentication
    r = redis.StrictRedis(
        host=host,
        port=port,
        username=username,
        password=password,
        ssl=True,  # Enable SSL/TLS
        decode_responses=True # Decode responses to strings
    )

    # Test the connection with a ping
    r.ping()
    print(f"Successfully connected to Redis at {host}:{port}!")
    print("Redis connection successful. You can now execute commands.")

    # Example: set and get a key
    r.set("test_key", "hello")
    value = r.get("test_key")
    print(f"Set 'test_key' to 'hello' and retrieved: {value}")

except redis.exceptions.ConnectionError as e:
    print(f"Could not connect to Redis. Error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
