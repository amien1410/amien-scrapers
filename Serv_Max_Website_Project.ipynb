```python
# @title
import os
import re
import requests
import trafilatura
from urllib.parse import urlparse
import shutil

# Folder to store extracted text
output_dir = "goulburn"
os.makedirs(output_dir, exist_ok=True)

# Name of the final zip file (without .zip extension)
zip_name = "goulburn"

def clean_filename(url):
    """Convert URL to a safe filename"""
    parsed = urlparse(url)
    filename = parsed.netloc + parsed.path
    filename = re.sub(r'[^a-zA-Z0-9_-]', '_', filename)  # replace unsafe chars
    return filename.strip("_") + ".txt"

for url in goulburn:
    try:
        print(f"[INFO] Fetching: {url}")
        response = requests.get(url, timeout=10)
        response.raise_for_status()

        # Extract main content with trafilatura
        extracted_text = trafilatura.extract(response.text)
        if not extracted_text:
            print(f"[WARN] Could not extract main content from {url}")
            continue

        # Create file name based on URL
        filename = clean_filename(url)
        filepath = os.path.join(output_dir, filename)

        # Save extracted text
        with open(filepath, "w", encoding="utf-8") as f:
            f.write(extracted_text)

        print(f"[OK] Saved: {filepath}")

    except requests.exceptions.RequestException as e:
        print(f"[ERROR] Failed to fetch {url} -> {e}")

# --- Zip the output folder ---
shutil.make_archive(zip_name, 'zip', output_dir)
print(f"\n[FINAL] All extracted pages have been zipped into: {zip_name}.zip")
```

```python
#@title pages extractor
import os
import re
import requests
import trafilatura
from bs4 import BeautifulSoup
from urllib.parse import urlparse
import shutil

# -----------------------------
# Helpers
# -----------------------------
def clean_filename(name):
    """Clean a string so it is safe as a filename."""
    name = re.sub(r'[^a-zA-Z0-9_-]', '_', name)
    name = re.sub(r'_+', '_', name)
    return name.strip("_")


def title_from_html(html):
    """Extract <title> and clean it."""
    soup = BeautifulSoup(html, "html.parser")
    title = soup.title.string if soup.title else "untitled_page"
    return clean_filename(title)


# -----------------------------
# Main Extraction Logic
# -----------------------------
output_dir = "sjpbowling3"
zip_name = "sjpbowling3"

os.makedirs(output_dir, exist_ok=True)

for url in urls:
    try:
        print(f"[INFO] Fetching: {url}")
        response = requests.get(url, timeout=10)
        response.raise_for_status()

        html_source = response.text

        # Extract title for both filenames
        title_clean = title_from_html(html_source)

        # -----------------------------
        # Save Raw HTML (title.html)
        # -----------------------------
        html_filename = f"{title_clean}.html"
        html_filepath = os.path.join(output_dir, html_filename)

        with open(html_filepath, "w", encoding="utf-8") as f:
            f.write(html_source)

        print(f"[HTML] Saved: {html_filepath}")

        # -----------------------------
        # Save Extracted Text (title.txt)
        # -----------------------------
        extracted_text = trafilatura.extract(html_source)
        if not extracted_text:
            print(f"[WARN] Could not extract main content from {url}")
            continue

        txt_filename = f"{title_clean}.txt"
        txt_filepath = os.path.join(output_dir, txt_filename)

        with open(txt_filepath, "w", encoding="utf-8") as f:
            f.write(f"{title_clean}\n\n")  # also add title inside the text
            f.write(extracted_text)

        print(f"[TEXT] Saved: {txt_filepath}")

    except requests.exceptions.RequestException as e:
        print(f"[ERROR] Failed to fetch {url} -> {e}")

# -----------------------------
# Zip the output directory
# -----------------------------
shutil.make_archive(zip_name, 'zip', output_dir)
print(f"\n[FINAL] All extracted files zipped into: {zip_name}.zip")
```

```python
import os

def merge_text_files(folder_path, output_file):
    """
    Merge all .txt files in the specified folder into a single text file.
    Each file's content is separated by a header (the filename).
    """
    # Get all text files in folder
    text_files = [f for f in os.listdir(folder_path) if f.endswith(".txt")]

    if not text_files:
        print("No text files found in the folder.")
        return

    with open(output_file, "w", encoding="utf-8") as outfile:
        for filename in text_files:
            file_path = os.path.join(folder_path, filename)

            # Write header
            header = f"\n\n===== {filename} =====\n\n"
            outfile.write(header)

            # Write file content
            with open(file_path, "r", encoding="utf-8") as infile:
                content = infile.read()
                outfile.write(content)

    print(f"‚úÖ Successfully merged {len(text_files)} files into '{output_file}'")

if __name__ == "__main__":
    # Example usage:
    folder = "sjpbowling3"          # Folder containing text files
    output = "sjpbowling3.txt" # Output file name
    merge_text_files(folder, output)
```

```python
#@title Stripdrains extractor (from local HTML files)
import os
import re
import trafilatura
import shutil

# ===== CONFIG =====
input_dir = "hallfn"
output_dir = "hallfn_text"
zip_name = "hallfn_text"
# ==================

def clean_filename(name):
    """Make sure filenames are safe and consistent"""
    name = re.sub(r'[^a-zA-Z0-9_-]', '_', name)
    return name.strip("_") + ".txt"

os.makedirs(output_dir, exist_ok=True)

# List all HTML files in folder
html_files = [f for f in os.listdir(input_dir) if f.endswith(".html")]

print(f"[INFO] Found {len(html_files)} HTML files to process.")

for file in html_files:
    input_path = os.path.join(input_dir, file)
    output_filename = clean_filename(os.path.splitext(file)[0])
    output_path = os.path.join(output_dir, output_filename)

    try:
        # Read HTML content
        with open(input_path, "r", encoding="utf-8") as f:
            html_text = f.read()

        # Extract main content with trafilatura
        extracted_text = trafilatura.extract(html_text)
        if not extracted_text:
            print(f"[WARN] Could not extract main content from {file}")
            continue

        # Save extracted text
        with open(output_path, "w", encoding="utf-8") as f:
            f.write(extracted_text)

        print(f"[OK] Saved: {output_filename}")

    except Exception as e:
        print(f"[ERROR] Failed to process {file} -> {e}")

# --- Zip the output folder ---
shutil.make_archive(zip_name, 'zip', output_dir)
print(f"\n[FINAL] All extracted text files have been zipped into: {zip_name}.zip")
```

```python
import os

def merge_text_files(folder_path, output_file):
    """
    Merge all .txt files in the specified folder into a single text file.
    Each file's content is separated by a header (the filename).
    """
    # Get all text files in folder
    text_files = [f for f in os.listdir(folder_path) if f.endswith(".txt")]

    if not text_files:
        print("No text files found in the folder.")
        return

    with open(output_file, "w", encoding="utf-8") as outfile:
        for filename in text_files:
            file_path = os.path.join(folder_path, filename)

            # Write header
            header = f"\n\n===== {filename} =====\n\n"
            outfile.write(header)

            # Write file content
            with open(file_path, "r", encoding="utf-8") as infile:
                content = infile.read()
                outfile.write(content)

    print(f"‚úÖ Successfully merged {len(text_files)} files into '{output_file}'")

if __name__ == "__main__":
    # Example usage:
    folder = "hallfn_text"          # Folder containing text files
    output = "hallfn.txt" # Output file name
    merge_text_files(folder, output)
```
