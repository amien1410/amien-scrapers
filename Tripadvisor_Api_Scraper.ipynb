```python
import requests

url = "https://api.content.tripadvisor.com/api/v1/location/search?"

headers = {
    "accept": "application/json"
}
params = {
    "key": "",
    "searchQuery": "banjarmasin",
    "category": "restaurant"
}

response = requests.get(url, headers=headers, params=params)

print(response.text)
```
