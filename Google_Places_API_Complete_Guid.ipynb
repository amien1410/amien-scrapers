```python
#@title Google Places API Modules
import sys
!{sys.executable} -m pip install google-maps-places

# Import all packages
from google.maps import places_v1
from google.geo.type.types import Viewport
from google.maps.places_v1 import types as places_v1_types
from google.type import latlng_pb2
from google.protobuf.json_format import MessageToDict, MessageToJson
import pandas as pd
import json
import math

# Load the API Key and create the Client.
API_Key = ""
client = places_v1.PlacesClient(
    client_options={"api_key": API_Key}
)
```

```python
#@title Text Search

# The address or place name to search for
address = "Starbucks Singapore"

# Create a SearchTextRequest object
# Note: The 'fields' parameter is NOT included here.
request = places_v1.SearchTextRequest(
    text_query=address
)
# Include the field mask
list_of_fields = ["id","displayName",
                  "formattedAddress",
                  "primaryType", "types",
                  "priceLevel","priceRange",
                  "rating", "userRatingCount",
                  "location"]
# fieldMask = "places.id,places.formattedAddress,places.displayName"
fieldMask = ""
for x in range(len(list_of_fields)):
    fieldMask += "places."
    fieldMask += f"{list_of_fields[x]}"
    if x != len(list_of_fields) - 1:
        fieldMask += ","

response = client.search_text(request=request, metadata=[("x-goog-fieldmask",fieldMask)])
display(response)
```

```python
#@title --- Define the location bias parameters ---
# Choose the Starbucks Compass One Latitude and Longitude
# --- Define the location restriction parameters ---
lat_used = 1.3923968999
lng_used = 103.8949150999

# Create the LatLng object for the center of the circle
center_point = latlng_pb2.LatLng(
    latitude=lat_used,
    longitude=lng_used
)

# Define a radius for the search restriction (in meters)
radius_meters = 100.0

# Create the Circle object
circle_area = places_v1.types.Circle(
    center=center_point,
    radius=radius_meters
)

# Correctly add the circle to the location restriction
# You must directly pass the Circle object to the LocationRestriction constructor.
location_bias = places_v1.SearchTextRequest.LocationBias(
    circle=circle_area
)
```

```python
#@title --- Define the location restriction parameters ---

# Approximate Earth's radius in meters for latitude
# The length of a degree of latitude is roughly constant
# A more accurate calculation would be needed for larger distances
EARTH_RADIUS_METERS = 6378137.0

# Calculate the change in latitude and longitude for the given radius
# One degree of latitude is approximately 111,132 meters.
delta_lat = radius_meters / 111132.0

# The length of a degree of longitude varies with latitude.
# At the equator, it's similar to a degree of latitude.
# At higher latitudes, it's shorter.
delta_lng = radius_meters / (111132.0 * math.cos(math.radians(lat_used)))

# Calculate the coordinates for the southwest (low) corner
low_lat = lat_used - delta_lat
low_lng = lng_used - delta_lng

# Calculate the coordinates for the northeast (high) corner
high_lat = lat_used + delta_lat
high_lng = lng_used + delta_lng

# Create the LatLng objects for the corners
low_point = latlng_pb2.LatLng(
    latitude=low_lat,
    longitude=low_lng
)
high_point = latlng_pb2.LatLng(
    latitude=high_lat,
    longitude=high_lng
)

# Create the Viewport object
viewport = Viewport(
    low=low_point,
    high=high_point
)

# Correctly add the rectangle to the location restriction
location_restriction = places_v1.SearchTextRequest.LocationRestriction(
    rectangle=viewport
)
```

```python
#@title Applying location restriction and bias on requests

# The address or place name to search for
address = "Starbucks Singapore"

# Create a SearchTextRequest object
# Note: The 'fields' parameter is NOT included here.
request = places_v1.SearchTextRequest(
    text_query=address,
    # location_bias=location_bias,
    location_restriction=location_restriction
)
# Include the field mask
list_of_fields = ["id","displayName",
                  "formattedAddress",
                  "primaryType", "types",
                  "priceLevel","priceRange",
                  "rating", "userRatingCount",
                  "location"]
# fieldMask = "places.id,places.formattedAddress,places.displayName"
fieldMask = ""
for x in range(len(list_of_fields)):
    fieldMask += "places."
    fieldMask += f"{list_of_fields[x]}"
    if x != len(list_of_fields) - 1:
        fieldMask += ","

response = client.search_text(request=request, metadata=[("x-goog-fieldmask",fieldMask)])
display(response)
```

```python
#@title Nearby Search Illustration

def generate_circle_centers(lat, lng, radius_km, circle_radius_m):
    R = circle_radius_m
    radius_m = radius_km * 1000
    big_radius = radius_m + R

    d = R * math.sqrt(3)
    h = 1.5 * R

    meters_per_degree_lat = 111320.0
    meters_per_degree_lng = meters_per_degree_lat * math.cos(math.radians(lat))

    i_min = math.floor(-big_radius / h)
    i_max = math.ceil(big_radius / h)

    centers = []

    for i in range(int(i_min), int(i_max) + 1):
        y = i * h
        j0 = d / 2.0 if i % 2 != 0 else 0.0

        j_min_val = (-big_radius - j0) / d
        j_max_val = (big_radius - j0) / d
        j_min = math.ceil(j_min_val)
        j_max = math.floor(j_max_val)

        for j in range(int(j_min), int(j_max) + 1):
            x = j0 + j * d
            if x*x + y*y <= big_radius * big_radius:
                delta_lat = y / meters_per_degree_lat
                delta_lng = x / meters_per_degree_lng
                new_lat = lat + delta_lat
                new_lng = lng + delta_lng
                centers.append((new_lat, new_lng))

    return centers

# Example usage
lat_used = 1.3923968999
lng_used = 103.8949150999
radius_km = 1
circle_radius_m = 100.0

centers = generate_circle_centers(lat_used, lng_used, radius_km, circle_radius_m)
print(f"Generated {len(centers)} circles.")

### To print the plot as shown above.
import matplotlib.pyplot as plt

# Split centers into latitudes/longitudes
lats = [c[0] for c in centers]
lons = [c[1] for c in centers]

plt.figure(figsize=(10, 8))
plt.scatter(lons, lats, s=5, alpha=0.7, c='blue')
plt.scatter(lng_used, lat_used, s=100, c='red', marker='*', label='Original Center')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title(f'Circle Centers (Total: {len(centers)})')
plt.grid(ls='--', alpha=0.5)
plt.legend()
plt.tight_layout()
# plt.savefig('circle_centers.png', dpi=150)
plt.show()
```
