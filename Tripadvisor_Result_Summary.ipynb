```python
import pandas as pd

file_path = '/content/drive/MyDrive/Upwork/patrick-bwn-project/TripAdvisor - State Based Search/tripadvisor_all_state_results.csv'
df = pd.read_csv(file_path)
display(df.head())
```

```python
print("Unique values per column:")
for column in df.columns:
    print(f"{column}: {df[column].nunique()} unique values")
```

```python
unique_titles = df['title'].nunique()
print(f"Number of unique titles: {unique_titles}")
```

```python
unique_primary_info = df['primary_info'].nunique()
print(f"Number of unique values in 'primary_info' column: {unique_primary_info}")
```

```python
theaters_count = df[df['primary_info'] == 'Theaters'].shape[0]
print(f"Number of entries with 'primary_info' as 'Theaters': {theaters_count}")

print("\nFirst 5 entries where 'primary_info' is 'Theaters':")
display(df[df['primary_info'] == 'Theaters'].head())
```
