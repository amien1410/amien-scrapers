```python
import re
from google.maps import places_v1
from google.api_core.exceptions import GoogleAPIError

def validate_api_key(API_KEY):
    try:
        client = places_v1.PlacesClient(
            client_options={"api_key": API_KEY}
        )
        address = "Starbucks Singapore"
        request = places_v1.SearchTextRequest(text_query=address)
        list_of_fields = ["id","displayName", "formattedAddress", "primaryType", "types", "priceLevel","priceRange", "rating", "userRatingCount", "location"]
        fieldMask = ""
        for x in range(len(list_of_fields)):
            fieldMask += "places."
            fieldMask += f"{list_of_fields[x]}"
            if x != len(list_of_fields) - 1:
                fieldMask += ","

        response = client.search_text(request=request, metadata=[("x-goog-fieldmask",fieldMask)])
        # If the call succeeds without raising an exception, it's considered valid.
        return True
    except GoogleAPIError as e:
        # print(f"API Key validation failed: {e}")
        return False
    except Exception as e:
        # print(f"An unexpected error occurred during API key validation: {e}")
        return False

def find_api_keys(text):
    pattern = r'\bAIzaSy[A-Za-z0-9_\-]{30,60}\b'
    api_keys = re.findall(pattern, text)
    return set(list(api_keys))
```

```python
# myText = r"""
# put your text contains google places api key here
# """

myText = r"""

"""

api_keys = find_api_keys(myText)
print(f"Found api keys: {len(api_keys)}")

for key in api_keys:
    is_valid = validate_api_key(key)
    if is_valid:
      print(key)
```
