```python
from openai import OpenAI

client = OpenAI(api_key="")
audio_file= open("/content/alloy.wav", "rb")

transcription = client.audio.transcriptions.create(
    model="gpt-4o-transcribe",
    file=audio_file
)

print(transcription.text)
```
