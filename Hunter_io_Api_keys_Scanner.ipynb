```python
import requests
import re
from typing import List

BASE_URL = "https://api.hunter.io/v2"

def validate_api_key(api_key: str) -> dict:
    url = f"{BASE_URL}/account"
    params = {"api_key": api_key}

    try:
        response = requests.get(url, params=params, timeout=10)
        response.raise_for_status()
        return {
            "valid": True,
            "data": response.json()
        }
    except requests.exceptions.HTTPError:
        return {
            "valid": False,
            "error": response.json()
        }
    except requests.exceptions.RequestException as e:
        return {
            "valid": False,
            "error": str(e)
        }

def extract_api_keys(text: str) -> List[str]:
    pattern = r"\b[a-z0-9]{40}\b"
    keys = re.findall(pattern, text)
    return list(set(keys))  # remove duplicates
```

```python
myText = """
api_keys
"""

api_keys = extract_api_keys(myText)

valid_keys = []
valid_count = 0
invalid_count = 0

for key in api_keys:
    result = validate_api_key(key)

    if result.get("valid"):
        valid_keys.append(key)
        valid_count += 1
        print(f"✅ VALID")
    else:
        invalid_count += 1
        print(f"❌ INVALID: {key}")

# Save valid keys to file
with open("hunter_api_keys.txt", "w") as f:
    for key in valid_keys:
        f.write(key + "\n")

print("\n=== SUMMARY ===")
print(f"Total keys found : {len(api_keys)}")
print(f"Valid keys       : {valid_count}")
print(f"Invalid keys     : {invalid_count}")
print("Saved valid keys to hunter_api_keys.txt")
```
