```python
import requests

url = "https://walmart2.p.rapidapi.com/searchV2"
querystring = {"query":"xiaomi"}
headers = {
    "x-rapidapi-host": "walmart2.p.rapidapi.com",
    'x-rapidapi-key': ''
}

response = requests.get(url, headers=headers, params=querystring)
response.json()
```
