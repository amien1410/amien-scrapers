```python
import aiohttp
from bs4 import BeautifulSoup
import asyncio

URL = "https://faces.pearlbrandsonline.com/shop/?orderby=menu_order&cpg_page=1"
```

Now, let's define an asynchronous function to fetch the HTML content of the page.

```python
async def fetch_html(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
        'Accept-Language': 'en-US,en;q=0.5',
        'Connection': 'keep-alive'
    }
    async with aiohttp.ClientSession(headers=headers) as session:
        async with session.get(url) as response:
            response.raise_for_status()  # Raise an exception for HTTP errors
            return await response.text()
```

Next, we will fetch the page, parse it with BeautifulSoup, and extract the product information. I'll search for `li.product` elements, and within each, look for the product title (`h2.woocommerce-loop-product__title`) and the link (`a.woocommerce-LoopProduct-link`).

```python
async def get_products_data(url):
    html = await fetch_html(url)
    soup = BeautifulSoup(html, 'html.parser')

    products = []
    # Find all product items. The website uses 'li.product' for each product entry.
    product_elements = soup.select('li.product')

    for product_elem in product_elements:
        name_elem = product_elem.select_one('h2.woocommerce-loop-product__title')
        url_elem = product_elem.select_one('a.woocommerce-LoopProduct-link')

        product_name = name_elem.get_text(strip=True) if name_elem else 'N/A'
        product_url = url_elem['href'] if url_elem and 'href' in url_elem.attrs else 'N/A'

        products.append({'name': product_name, 'url': product_url})

    return products

# Run the asynchronous function
product_list = await get_products_data(URL)

# Display the extracted product information
for product in product_list:
    print(f"Name: {product['name']}, URL: {product['url']}")
```
