```python
import re
from contextlib import suppress
from dataclasses import dataclass
from typing import Optional
from pprint import pprint

from scrapfly import ScrapeConfig, ScrapflyClient, ScrapeApiResponse
from bs4 import BeautifulSoup

def validate_scrapfly_api_key(api_key: str) -> bool:
    """Validates a Scrapfly API key by attempting a basic scrape."""
    try:
        client = ScrapflyClient(key=api_key)
        # Attempt a lightweight scrape to validate the key
        response: ScrapeApiResponse = client.scrape(
            scrape_config=ScrapeConfig(url='https://news.ycombinator.com/', asp=False, country='US')
        )
        # Check if the scrape was successful (e.g., no errors in response, status code 200)
        if response.success:
            # print("Scrapfly API key is valid!")
            return True
        else:
            # print(f"Scrapfly API key validation failed: {response.error}")
            return False
    except Exception as e:
        # print(f"An error occurred during API key validation: {e}")
        return False

# Example usage of the validation function:
# Replace 'YOUR_API_KEY' with your actual key to test.
# is_valid = validate_scrapfly_api_key(api_key='scp-live-8afea162f0f44a54a71193fb1e19cae6')
# print(f"API Key valid: {is_valid}")
```
