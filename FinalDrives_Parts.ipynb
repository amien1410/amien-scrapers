```python
import requests
```

```python
# get all makes
url = "https://quotesv2.finaldrives.com/api/data/machineMakes"
res = requests.get(url)
data = res.json()
make_data = [
    {
        'Id': item['Id'],
        'Make': item['Make'],
        'ModelComponentTypes': item['ModelComponentTypes']
    }
    for item in data['Items']
]
make_data
```

```python
url2 = "https://quotesv2.finaldrives.com/api/data/machineModels/1?componentType=FinalDrive"
res = requests.get(url2)
data = res.json()
data['Items']
```

```python
all_models = []
component_type = "FinalDrive"  # Define the component type you need

for make in make_data:
    print(f"{make['Make']} : {make['Id']}")
    make_id = make['Id']
    url2 = f"https://quotesv2.finaldrives.com/api/data/machineModels/{make_id}?componentType={component_type}"

    try:
        res = requests.get(url2)
        res.raise_for_status()  # Check for HTTP errors
        model_data = res.json()

        # Check if 'Items' exists and is non-empty
        if 'Items' in model_data and model_data['Items']:
            # Store model names and IDs by make ID
            for model in model_data['Items']:
                all_models.append({
                    'ModelId': model['Id'],
                    'ModelName': model['Model'],
                    'MakeName': make['Make']
                })
        else:
            print(f"No models found for make '{make['Make']}' with ID {make_id}")

    except requests.exceptions.RequestException as e:
        print(f"Failed to fetch models for make '{make['Make']}' with ID {make_id}: {e}")
```

```python
import pandas as pd
df = pd.DataFrame(all_models)
df.to_csv('all_models.csv', index=False)
df
# len(all_models)
# all_models
```

```python
import random

def generate_random_headers():
    user_agent_base = "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:{version}) Gecko/20100101 Firefox/{version}"
    firefox_version = f"{random.randint(100, 150)}.0"

    headers = {
        "Host": "quotesv2.finaldrives.com",
        "User-Agent": user_agent_base.format(version=firefox_version),
        "Accept": "application/json, text/plain, */*",
        "Accept-Language": random.choice(["en-US,en;q=0.5", "en-GB,en;q=0.5", "en;q=0.5"]),
        "Accept-Encoding": random.choice(["gzip, deflate, br", "gzip, deflate, br, zstd"]),
        "Connection": "keep-alive",
        "Referer": "https://quotesv2.finaldrives.com/PP-Website/FinalDrive/lookup",
        "Cookie": f"_ga_F34SKVVXKK=GS1.1.{random.randint(1000000000, 2000000000)}.1.0.{random.randint(1000000000, 2000000000)}.0.0.0; _ga=GA1.1.{random.randint(100000000, 999999999)}.{random.randint(1000000000, 2000000000)}",
        "Sec-Fetch-Dest": "empty",
        "Sec-Fetch-Mode": "cors",
        "Sec-Fetch-Site": "same-origin",
        "Priority": "u=0"
    }

    return headers

# Example usage:
# random_headers = generate_random_headers()
# print(random_headers)
```

```python
# Function to fetch machines by model ID
def fetch_machines_by_model_id(model_id):
    url = f"https://quotesv2.finaldrives.com/api/data/getMachines/{model_id}?componentType=FinalDrive&language=en_US&recaptchaResponse=NA&source=PP-Website"
    headers = generate_random_headers()

    try:
        res = requests.get(url, headers=headers)
        res.raise_for_status()  # Check for HTTP errors
        data = res.json()

        # Check if 'Items' exists and is non-empty
        if 'Items' in data and data['Items']:
            return data['Items']
        else:
            print(f"No machines found for model ID {model_id}")
            return []

    except requests.exceptions.RequestException as e:
        print(f"Failed to fetch machines for model ID {model_id}: {e}")
        return []

# Loop through all models and fetch machine data
all_machine_data = []

for model in all_models:
    model_id = model['ModelId']
    model_name = model['ModelName']
    make_name = model['MakeName']

    print(f"Fetching machines for Model: {model_name} (ID: {model_id}), Make: {make_name}")
    machines = fetch_machines_by_model_id(model_id)
    print(machines)

    # Append the result with additional model and make information
    all_machine_data.append({
        'ModelId': model_id,
        'ModelName': model_name,
        'MakeName': make_name,
        'Machines': machines
    })
```
