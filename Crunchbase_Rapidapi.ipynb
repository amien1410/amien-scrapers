```python
import pprint
import requests

url = "https://crunchbase4.p.rapidapi.com/company"

payload = { "company_domain": "arcare.com.au" }
headers = {
	"x-rapidapi-key": "",
	"x-rapidapi-host": "crunchbase4.p.rapidapi.com",
	"Content-Type": "application/json"
}

response = requests.post(url, json=payload, headers=headers)

pprint.pprint(response.json())
```

```python
import requests

url = "https://website-intelligence.p.rapidapi.com/lookup"

querystring = {"domain":"microsoft.com","query":"microsoft"}

headers = {
	"x-rapidapi-key": "",
	"x-rapidapi-host": "website-intelligence.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```
