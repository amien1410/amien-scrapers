```python
import re
import requests

FOURSQUARE_BASE_URL = "https://api.foursquare.com/v3/places/search"
FOURSQUARE_PHOTOS_URL = "https://api.foursquare.com/v3/places/{fsq_id}/photos"
```

```python
def validate_api_key(api_key):
    headers = {
        "Accept": "application/json",
        "Authorization": api_key
    }
    params = {"query": "restaurant", "near": "Barcelona", "limit": 50}
    try:
        response = requests.get(FOURSQUARE_BASE_URL, headers=headers, params=params)
        if response.status_code == 200:
            return True
        else:
            return False
    except requests.exceptions.RequestException as e:
        return False

def find_api_keys(text):
    pattern = r'fsq3[A-Za-z0-9+/]{20,}='
    api_keys = re.findall(pattern, text)
    return set(list(api_keys))
```

```python
myText = r"""

"""

api_keys = find_api_keys(myText)
print(f"Found api keys: {len(api_keys)}")

for key in api_keys:
    is_valid = validate_api_key(key)
    if is_valid:
        print(key)
```
