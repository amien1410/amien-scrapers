```python
import os
import re
import requests
import trafilatura
from urllib.parse import urlparse
from bs4 import BeautifulSoup

def clean_filename(url):
    """Convert URL to a safe filename"""
    parsed = urlparse(url)
    filename = parsed.netloc + parsed.path
    filename = re.sub(r'[^a-zA-Z0-9_-]', '_', filename)  # replace unsafe chars
    return filename.strip("_") + ".txt"

def extract_text_with_title(url, html_text, output_dir):
    """Extract readable text and prepend the page title as header"""
    # Parse title using BeautifulSoup
    soup = BeautifulSoup(html_text, "html.parser")
    title = soup.title.string.strip() if soup.title and soup.title.string else "No Title Found"

    # Extract main readable text
    extracted_text = trafilatura.extract(html_text)
    if not extracted_text:
        print(f"[WARN] Could not extract main content from {url}")
        extracted_text = ""

    # Prepare full text (Title + separator + content)
    final_text = f"{title}\n{'=' * len(title)}\n\n{extracted_text}"

    # Create file name based on URL
    filename = clean_filename(url)
    filepath = os.path.join(output_dir, filename)

    # Save extracted text
    with open(filepath, "w", encoding="utf-8") as f:
        f.write(final_text)

    print(f"[OK] Saved: {filepath}")

# === Main Config ===
output_dir = "secondhandforklifts"

os.makedirs(output_dir, exist_ok=True)

# Assume unique_urls is defined elsewhere
for url in unique_urls:
    try:
        print(f"[INFO] Fetching: {url}")
        response = requests.get(url, timeout=10)
        response.raise_for_status()

        # Extract and save with title
        extract_text_with_title(url, response.text, output_dir)

    except requests.exceptions.RequestException as e:
        print(f"[ERROR] Failed to fetch {url} -> {e}")
```

```python
import os

def merge_text_files(folder_path, output_file):
    """
    Merge all .txt files in the specified folder into a single text file.
    Each file's content is preceded by a numbered header (the filename),
    and a blank line is inserted before each new file's section.
    """
    # Get all text files in folder
    text_files = [f for f in os.listdir(folder_path) if f.endswith(".txt")]
    text_files.sort()  # Sort alphabetically for consistent order

    if not text_files:
        print("No text files found in the folder.")
        return

    with open(output_file, "w", encoding="utf-8") as outfile:
        for i, filename in enumerate(text_files, start=1):
            file_path = os.path.join(folder_path, filename)

            # Read file content
            with open(file_path, "r", encoding="utf-8") as infile:
                content = infile.read().strip()

            # Write header and content with spacing
            outfile.write(f"\n=== {i}. {filename.replace('.txt', '')} ===\n")
            outfile.write(content + "\n")

    print(f"✅ Successfully merged {len(text_files)} files into '{output_file}'")

if __name__ == "__main__":
    # Example usage:
    folder = "secondhandforklifts"          # Folder containing text files
    output = "secondhandforklifts2.txt"      # Output file name
    merge_text_files(folder, output)
```
