```python
import requests

CRUNCHBASE_API_KEY = "YOUR_API_KEY_HERE"
BASE_URL = "https://api.crunchbase.com/api/v4"

HEADERS = {
    "X-cb-user-key": CRUNCHBASE_API_KEY,
    "Content-Type": "application/json"
}

def search_company_by_name(company_name, limit=10):
    """
    Search Crunchbase organizations by company name
    """
    url = f"{BASE_URL}/searches/organizations"

    payload = {
        "query": [
            {
                "field_id": "name",
                "operator_id": "contains",
                "values": [company_name]
            }
        ],
        "limit": limit
    }

    response = requests.post(url, headers=HEADERS, json=payload)
    response.raise_for_status()

    data = response.json()

    companies = []
    for item in data.get("entities", []):
        props = item.get("properties", {})
        companies.append({
            "uuid": item.get("uuid"),
            "name": props.get("name"),
            "website": props.get("website_url"),
            "short_description": props.get("short_description"),
            "country": props.get("country_code")
        })

    return companies


if __name__ == "__main__":
    results = search_company_by_name("OpenAI", limit=5)
    for r in results:
        print(r)
```
