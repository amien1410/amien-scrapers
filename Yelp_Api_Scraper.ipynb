```python
import re
import requests

url = "https://api.yelp.com/v3/businesses/search?location=san%20fransisco&term=restaurants&sort_by=best_match&limit=50"

def validate_api_key(api_key):
    headers = {
        "accept": "application/json",
        "authorization": f"Bearer {api_key}"
    }

    response = requests.get(url, headers=headers)

    if response.json() and "businesses" in response.json():
        return True

    else:
        return False

def find_api_keys(text):
    pattern = r'\b[A-Za-z0-9_-]+Yx\b'
    api_keys = re.findall(pattern, text)
    return set(list(api_keys))
```
