```python
from murf import Murf

myText = "Murf offers a variety of voice IDs for different languages and accents. You can choose a voice ID that best suits your application’s needs. You can see the full list of available voice IDs here, or fetch the list programmatically using the list_voices endpoint"

client = Murf(api_key="")

audio_res = client.text_to_speech.generate(
    text=myText,
    voice_id="en-US-terrell",
)

print(audio_res.audio_file)
```
