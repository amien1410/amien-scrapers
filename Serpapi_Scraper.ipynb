```python
import re
from serpapi import GoogleSearch

def validate_api_key(api_key: str) -> bool:
  """Validates the SerpApi API key by performing a simple search."""
  try:
    search = GoogleSearch({
        "q": "coffee",
        "location": "Austin,Texas",
        "api_key": api_key
    })
    search.get_dict()
    return True
  except Exception as e:
    print(f"API Key validation failed: {e}")
    return False

def find_api_keys(text):
  pattern = r'(?<![A-Fa-f0-9])[a-f0-9]{64}(?![A-Fa-f0-9])'
  api_keys = re.findall(pattern, text)
  return set(list(api_keys))
```

```python
myText = r"""

"""

api_keys = find_api_keys(myText)
print(f"Found api keys: {len(api_keys)}")

for key in api_keys:
    is_valid = validate_api_key(key)
    if is_valid:
      print(key)
```
