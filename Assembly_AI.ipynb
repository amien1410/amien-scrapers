```python
def validate_api_key(api_key: str) -> bool:
    try:
        aai.settings.api_key = api_key
        transcriber = aai.Transcriber()

        transcript = transcriber.transcribe(
            "https://storage.googleapis.com/aai-web-samples/gettysburg.wav"
        )

        # If API key is valid, transcript.text will exist
        if transcript.text:
            return True
        return False

    except Exception:
        # Any error means invalid key (401 Unauthorized, etc.)
        return False
```

```python
import re

text = """

"""

# Regex pattern:
# Matches: ASSEMBLYAI_API_KEY= followed by 32 hex characters
pattern = r"ASSEMBLYAI_API_KEY=([a-fA-F0-9]{32})"

keys = re.findall(pattern, text)

# Remove duplicates
keys = list(set(keys))
print("\nTotal keys found:", len(keys))

valid_keys = []
invalid_keys = []

for key in keys:
    print(f"Validating: {key}")

    if validate_api_key(key):
        print(" → VALID")
        valid_keys.append(key)
    else:
        print(" → INVALID")
        invalid_keys.append(key)

print("\n==================================================")
print("VALID KEYS:")
for k in valid_keys:
    print(k)

print("\nINVALID KEYS:")
for k in invalid_keys:
    print(" -", k)
print("==================================================\n")
```

```python
import re

text = r"""

"""

# Regex:
# 1. Capture os.getenv("ASSEMBLYAI_API_KEY", "<default_value>")
# 2. Then extract only real API key patterns

# Extract all default values provided in os.getenv("ASSEMBLYAI_API_KEY", "<HERE>")
# pattern = r'os\.getenv\(\s*"ASSEMBLYAI_API_KEY"\s*,\s*"([^"]*)"\s*\)'
pattern = r'aai\.settings\.api_key\s*=\s*"([a-fA-F0-9]{32})"'

raw_values = re.findall(pattern, text)

# Now filter for REAL AssemblyAI API keys:
# 32 hex or ap2_UUID patterns
real_key_pattern = r'^(?:[a-fA-F0-9]{32}|ap2_[A-Za-z0-9\-]{36,})$'

filtered_keys = [
    val for val in raw_values
    if re.match(real_key_pattern, val)
]

# Remove duplicates
filtered_keys = list(set(filtered_keys))

# print("Extracted AssemblyAI API Keys:")
# for k in filtered_keys:
#     print(" -", k)

print("\nTotal:", len(filtered_keys))

valid_keys = []
invalid_keys = []

for key in filtered_keys:
    print(f"Validating: {key}")

    if validate_api_key(key):
        print(" → VALID")
        valid_keys.append(key)
    else:
        print(" → INVALID")
        invalid_keys.append(key)

print("\n==================================================")
print("VALID KEYS:")
for k in valid_keys:
    print(k)

print("\nINVALID KEYS:")
for k in invalid_keys:
    print(" -", k)
print("==================================================\n")
```
