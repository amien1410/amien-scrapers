```python
import re
from supadata import Supadata, SupadataError

def validate_api_key(api_key: str) -> bool:
    try:
        client = Supadata(api_key=api_key)
        transcript = client.transcript(
          url="https://x.com/SpaceX/status/1481651037291225113",
          lang="en",  # Optional: preferred language
          text=True,  # Optional: return plain text instead of timestamped chunks
          mode="auto"  # Optional: "native", "auto", or "generate"
        )
        return True
    except SupadataError:
        return False
    except Exception:
        return False


def extract_api_keys(text: str) -> list[str]:
    pattern = r"\bsd_[a-zA-Z0-9]{32,40}\b"
    return set(list(re.findall(pattern, text)))
```

```python
import time

myText = r"""


"""

keys = extract_api_keys(myText)
print(f"Found: {len(keys)} keys")

for key in keys:
    is_valid = validate_api_key(key)
    if is_valid:
        print(key)
    time.sleep(2)
```
