```python
data = []

# Find all rows with class "site-search-row"
for tr in soup.find_all("tr", class_="site-search-row"):
    cells = [td.get_text(strip=True) for td in tr.find_all("td")]
    data.append(cells)

# Print extracted rows
for row in data:
    print(row)
```

```python
# Find the table
table = soup2.find("table", class_="detailTable")

# Find all <th> and their corresponding <td>
data = {}
ths = table.find_all("th")
for th in ths:
    label = th.get_text(strip=True)
    td = th.find_next_sibling("td")   # Get the next <td> after the <th>
    if td:
        # Replace <br> with newlines, then strip
        value = td.decode_contents().replace("<br>", "\n")
        clean_value = BeautifulSoup(value, "html.parser").get_text().strip()
        data[label] = clean_value

# Extract results
owner = data.get("Owner").strip()
taxpayer = data.get("Taxpayer").strip()

print("Owner:\n", owner)
print("Taxpayer:\n", taxpayer)
```
